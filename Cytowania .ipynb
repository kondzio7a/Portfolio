```python
import time
import random
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from fake_useragent import UserAgent
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()
ua = UserAgent()
options.add_argument(f"user-agent={ua.random}")
options.add_argument("--headless")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--incognito")
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

def simulate_user_interaction(driver):
    scroll_distance = random.randint(100, 1000)
    driver.execute_script(f"window.scrollBy(0, {scroll_distance});")
    time.sleep(random.uniform(0.5, 2))
    if random.choice([True, False]):
        elements = driver.find_elements(By.CSS_SELECTOR, 'a')
        if elements:
            random_element = random.choice(elements)
            if random_element.is_displayed():
                driver.execute_script("arguments[0].scrollIntoView();", random_element)
                ActionChains(driver).move_to_element(random_element).click().perform()
                print("Kliknięto na element.")
    else:
        hover_elements = driver.find_elements(By.CSS_SELECTOR, 'a')
        if hover_elements:
            random_hover_element = random.choice(hover_elements)
            if random_hover_element.is_displayed():
                driver.execute_script("arguments[0].scrollIntoView();", random_hover_element)
                ActionChains(driver).move_to_element(random_hover_element).perform()
                print(f"Najechano na element: {random_hover_element.text}")
patent_numbers = ["US1234567A", "EP7654321B1", "WO2023123456"]  
results_df = pd.DataFrame(columns=["patent", "patent_family"])
pattern = re.compile(r'([A-Z]{2}\d{6,}\w*)(?=\s*[*†])')

for patent_number in patent_numbers:
    url = "https://patents.google.com/"
    driver.get(url)
    simulate_user_interaction(driver)
    time.sleep(2)
    search_box = driver.find_element(By.NAME, "q")
    search_box.send_keys(patent_number)
    search_box.send_keys(Keys.RETURN)
    simulate_user_interaction(driver)
    time.sleep(3)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")
    full_text = soup.get_text(separator="\n", strip=True)
    matching_codes = pattern.findall(full_text)
    for code in matching_codes:
        results_df = pd.concat([results_df, pd.DataFrame([[patent_number, code]], columns=["patent", "patent_family"])])

print("Wyniki wyszukiwania patentów i ich rodzin:")
print(results_df)
results_df.to_csv('D:/AA Praktyki/patent_results.csv', index=False)
print("\nWszystkie wyniki zostały zapisane.")
driver.quit()
```

```python
import time
import random
import re
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from fake_useragent import UserAgent
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()
ua = UserAgent()
options.add_argument(f"user-agent={ua.random}")
options.add_argument("--headless") 
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--incognito")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
def simulate_user_interaction(driver):
    scroll_distance = random.randint(100, 1000)
    driver.execute_script(f"window.scrollBy(0, {scroll_distance});")
    time.sleep(random.uniform(0.5, 2))

def format_patent_number(patent_number):
    if patent_number.startswith("P."):  
        return patent_number.replace("P.", "PL")  
    return patent_number  
pattern = re.compile(r'([A-Z]{2}\d{6,}\w*)(?=\s*[*†])')

results_df = pd.DataFrame(columns=["patent", "patent_family"])
save_interval = 10 
output_file = "D:/AA Praktyki/patent_results2.csv"  


for index, patent_number in enumerate(array):
    patent_number = format_patent_number(patent_number)
    try:
   
        url = "https://patents.google.com/"
        driver.get(url)
        simulate_user_interaction(driver)
        time.sleep(5)  

        search_box = driver.find_element(By.NAME, "q")
        search_box.clear()  
        search_box.send_keys(patent_number) 
        search_box.send_keys(Keys.RETURN)
        simulate_user_interaction(driver)
        time.sleep(5) 
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, "html.parser")
        full_text = soup.get_text(separator="\n", strip=True)

        matching_codes = pattern.findall(full_text)

        for code in matching_codes:
            results_df = pd.concat([results_df, pd.DataFrame([[patent_number, code]], columns=["patent", "patent_family"])], ignore_index=True)

        if (index + 1) % save_interval == 0:
            results_df.to_csv(output_file, index=False)
            print(f"Zapisano wyniki po przetworzeniu {index + 1} numerów patentów.")

    except Exception as e:
        print(f"Błąd podczas przetwarzania numeru {patent_number}: {e}")

results_df.to_csv(output_file, index=False)
print("\nWszystkie wyniki zostały zapisane.")

driver.quit()
```
