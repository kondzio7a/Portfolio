```python
import os
import re
import pandas as pd
import pdfplumber
from docx import Document

def read_pdf(file_path):
    try:
        with pdfplumber.open(file_path) as pdf:
            text = ''
            for page in pdf.pages:
                text += page.extract_text()
        return text, pdf
    except Exception as e:
        print(f"Nie udao si odczyta pliku PDF: {file_path}. Bd: {e}")
        return "", None

def read_txt(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
        return text
    except Exception as e:
        print(f"Nie udao si odczyta pliku TXT: {file_path}. Bd: {e}")
        return ""

def read_docx(file_path):
    try:
        doc = Document(file_path)
        text = ''
        for paragraph in doc.paragraphs:
            text += paragraph.text + '\n'
        return text
    except Exception as e:
        print(f"Nie udao si odczyta pliku DOCX: {file_path}. Bd: {e}")
        return ""

def combine_hyphenated_words(text):
    return re.sub(r'(\w+)-\n(\w+)', r'\1\2', text)

def clean_text(text):
    cleaned_lines = []
    skip_section = False
    for line in text.split('\n'):
        if re.search(r'\bid:.*?podpisany\b', line, flags=re.IGNORECASE):
            skip_section = True
        elif line.strip() == '':
            skip_section = False
        if not skip_section and not re.match(r'^[\s\S]*{10,}[\s\S]*$', line):
            cleaned_lines.append(line)
    return '\n'.join(cleaned_lines)

def extract_section(text, start_keywords, end_keywords):
    start_index = -1
    for keyword in start_keywords:
        index = text.lower().find(keyword.lower())
        if index != -1:
            start_index = index
            break
    
    if start_index == -1:
        return None

    end_index = len(text)
    for end_keyword in end_keywords:
        idx = text.lower().find(end_keyword.lower(), start_index)
        if idx != -1 and idx < end_index:
            end_index = idx

    if start_index != -1 and end_index != -1:
        extracted_text = text[start_index:end_index]
        extracted_text = clean_text(extracted_text)
        extracted_text = re.sub(r'\b[0-9a-fA-F]{8}-[0-9a-fA-F]{4}-[0-9a-fA-F]{4}-[0-9a-fA-F]{4}-[0-9a-fA-F]{12}\b', '', extracted_text)
        return extracted_text
    return None

def find_sentences_with_keywords(text, keywords):
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', text)
    matching_sentences = [sentence for sentence in sentences if any(keyword in sentence.lower() for keyword in keywords)]
    return matching_sentences

def process_text_with_pages(text, keywords):
    sentences = re.split(r'\n+', text)
    combined_results = []

    i = 0
    while i < len(sentences):
        sentence = sentences[i].strip()
        current_page = None

        current_page_match = re.search(r'\b(\d+)\s*$', sentence)
        if current_page_match:
            current_page = int(current_page_match.group(1))
            if current_page > 1000:
                i += 1
                continue 
        else:
            i += 1
            continue

        next_page = None
        j = i + 1
        while j < len(sentences):
            next_sentence = sentences[j].strip()
            next_page_match = re.search(r'\b(\d+)\s*$', next_sentence)
            if next_page_match:
                next_page = int(next_page_match.group(1))
                if next_page >= current_page and next_page < 1000:
                    break  
                else:
                    next_page = None 
            j += 1

        if any(keyword in sentence.lower() for keyword in keywords):
            if next_page:
                combined_results.append((sentence, current_page, next_page))
            else:
                combined_results.append((sentence, current_page, current_page))
        
        i = j 
    
    return combined_results

def extract_pages(pdf, page_ranges):
    extracted_text = ''
    for start_page, end_page in page_ranges:
        for page_num in range(start_page, end_page + 1):
            try:
                extracted_text += pdf.pages[page_num - 1].extract_text() + '\n'
            except IndexError:
                print(f"Strona {page_num} nie istnieje w dokumencie PDF.")
    return extracted_text

def process_single_file(file_path):
    if file_path.endswith('.pdf'):
        text, pdf = read_pdf(file_path)
    elif file_path.endswith('.txt'):
        text = read_txt(file_path)
        pdf = None
    elif file_path.endswith('.docx'):
        text = read_docx(file_path)
        pdf = None
    else:
        print(f"Nieobsugiwany format pliku: {file_path}")
        return None

    if text:
        text = combine_hyphenated_words(text)

        start_keywords = ['spis treci', 'zawarto', 'spis zawartoci']
        end_keywords = [
            'zaczniki', 'zacznik', 'spis tabel','Spis tabel', 'spis obrazw', 'spis map',
            'strategia rozwoju gminy i miasta', 'Spis wykresw','strategia rozwoju miasta i gminy',
            'spis map i tabel', 'spis rysunkw i tabel', 'strategia rozwoju miasta',
            'spis ilustracji', 'wykaz tabel', 'opis konsultacji spoecznych',
            'perspektyw do roku 2030', 'system wdraania strategii rozwoju gminy wieruszw',
            'wieloletni plan inwestycyjny', 'srotegorotwju mita gminy tonowo pomorskie olet',
            '30strategia rozwoju gminy i miasta gryfw lski', 'wdraanie strategii',
            'spistabei rysunkw! wykresw', 'spjno z krajowymi i regionalnymi dokumentami strategicznymi',
            'system wdraania, monitorowania i ramy finansowe','emce re nie owe w. zbir nowak com (panom sa. za)',
            '3Strategia Rozwoju Gminy Bolesawiec na lata 2023- 2028','Spis map, tabel, wykresw i zdj',
            'Spis map, tabel, wykresw i zdj','5Program Rozwoju Gminy Miejskiej Kowary na lata 2021-2030',
            '1STRATEGIA ROZWOJU GMINY KUDOWA-ZDRJ NA LATA 2021-2027','Strategia rozwoju Gminy Miejskiej Nowa Ruda na lata 2014-2020',
            'Strategii Rozwoju Gminy Polanica-Zdrj','Spis map, tabel, wykresw i zdj UM A','ZESP DS. OPRACOWANIA STRATEGII ROZWOJU MIASTA WIEBODZICE',
            'Zesp ds. opracowania Strategii rozwoju Gminy Guszyca','4 5PRZEDSIBIORCZY WROCAW','SPIS RYSUNKW','Strategia rozwoju jest dokumentem planistycznych, o charakterze dugofalowym, okrelajcym kierunki i',
            'Regionalne Centrum Przedsibiorczoci','2Strategia Rozwoju Gminy Janikowo','2Strategia Rozwoju Gminy Janikowo na lata 2014-2020',
            'mgr Adam Staczyk','2Czas na Nowe! Strategia Rozwoju Gminy Nowe na lata 2012-2022','Oddajemy w Pastwa rce Strategi Rozwoju Gminy wiecie do 2030 roku.',
            'Zaoenia rozwoju powiatu tucholskiego','OBSZAR I CZAS REALIZACJI PLANU ROZWOJU LOKALNEGO','Chc podkreli, e Strategia Rozwoju Bydgoszczy do 2030 roku','Spis wykresw','Miasto Rejowiec Fabryczny',
            '- 1 -Strategia Rozwoju Gminy Siedliszcze','Strategia Rozwoju Gminy Modliborzyce','"Strategia Rozwoju Gminy Modliborzyce na lata 2020-2025','Strategia Rozwoju Gminy Modliborzyce na lata 2020-2025 w postaci dokumentu, to okrelenie',
            '"Spis wykresw, tabel, rysunkw','EKO-GEO Consulting Marek Biyk','3Strategia rozwoju Gminy Poniatowa na lata 2015-2025',
            '2Strategia rozwoju Gminy aszczw na lata 2007  2020','Spjno Strategii Lublin 2030 z dokumentami wyszego rzdu','Agencja Rozwoju Regionalnego S.A. w Gorzowie Wielkopolskim',
            'Zawarto programu interwencji w ramach','Zarzdzanie strategiczne jednostk samorzdu terytorialnego jest','Niniejszy dokument zosta stworzony w celu przedstawienia kierunkw rozwoju Gminy Ono Lubuskie',
            'Stron a 2Strategia rozwoju gminy wiebodzin 2030','Strategia rozwoju gminy Sulechw 2030 (zwana dalej strategi) stanowi kluczowy dokument',
            'Strategia rozwoju gminy Szprotawa na lata 2015- 2023 jest najwaniejszym dokumentem','15. Spis rysunkw, tabel i wykresw','Strategia Rozwoju Gminy to dokument, ktry w jasny sposb wskazuje kierunki',
            'Gwnym motywem opracowania Strategii Rozwoju gminy Rzgw bya potrzeba stworzenia','Strategia Rozwoju Powiatu Poddbickiego 2020 jest','2MMIIAASSTTOO II GGMMIINNAA PPRRZZEEDDBBRRZZ',
            '4STRATEGIA ROZWOJU GMINY WIELU 2025','1Strategia Rozwoju Gminy Miasta Bochnia na lata 2022-2030','Strategia Rozwoju Gminy Alwernia na lata 2022-2030, jest dokumentem, ktry okrela',
            '3Strategia Rozwoju Gminy Trzebinia na lata 2014-2020','4STRATEGIA ROZWOJU GMINY SZCZUCIN 2014 - 2020',
            'Strategia Rozwoju Gminy witniki Grne na lata 2015-2020+ 2Wprowadzenie','2Strategia Rozwoju Gminy Czarny Dunajec na lata 2021-2030',
            '3Strategia Rozwoju Gminy Wolbrom','System wdraania Strategii Rozwoju Gminy Zator na lata 2014-2022',
            'STRATEGIA ROZWOJU GMINY KOSZYCE NA LATA 20212026','GMINY NOWE BRZESKOSTRATEGIA ROZWOJU GMINY NOWE BRZESKO',
            'Rysunek','Strategia Rozwoju Gminy Radw na lata 2021-2030 1Wprowadzenie','"Projekt: Nowoczesne Zarzdzanie Publiczne o numerze POKL.05.02.01-00-025/12, realizowany w ramach Programu"',
            '3C I. W','Aktualne dokumenty SZJ znajduj si w sieci komputerowej w portalu pod adresem http://qsystem.nowysacz.pl Dokumenty w wersji 3',
            'Strona 4 z 99Strategia Rozwoju Gminy Wymierzyce na lata 2021-2030','Strona 5STRATEGIA ROZWOJU GMINY ELECHW NA LATA 2014  2020',
            'Strategia Rozwoju Gminy Miasta Gostynina jest dokumentem zawierajcym dugookresowy','Gminy Warka na lata 2005-2013  jest sformuowanie syntetycznego dokumentu',
            'Strategia Rozwoju Gminy Kozienice na lata 2018-2030jest kierunkowo-programowym planem','Padziernik 2007 2Zaktualizowana Strategia Rozwoju Gminy Miejskiej Legionowo do 2015 roku',
            '2Strategia Rozwoju Gminy Solec nad Wis na lata 2007 - 2020','SPIS MAP','Strategia Zrwnowaonego Rozwoju Gminy Halinw na lata 2016-2025',
            'Plan Rozwoju Lokalnego to dokument, ktry przedstaw ia zintegrowane i kompleksowe dzia ania','STRATEGIA ZRWNOWAONEGO ROZWOJU MIASTA I GMINY ZAKROCZYM DO 2020 ROKU',
            '3STRATEGIA ROZWOJU SPOECZNO-GOSPODARCZEGO MIASTA OTWOCKA','3STRATEGIA ROZWOJU GMINY SOCHOCIN NA LATA 2016-2020 z perspektyw do roku 2025',
            '3Strategia Zrwnowaonego Rozwoju Miasta Piastowa do 2020 roku','Strona | 1STRATEGIA ROZWOJU GMINY MIASTO PIONKI NA LATA 2016-2022',
            'Spis tablic','W myl rozwiza ustawowych Wadze Gminy omianki! Zobowizane s do dziaa na','E O2OCHAOC 4724477 UE FTSWORZKL8. Podpeay son',
            'Przesanki aktualizacji Strategii Rozwoju Gminy Zwole','Rozwj regionalny i lokalny ma szczeglne znaczenie dla procesw wzrostu gospodarczego',
            'Opracowanie: JODKA Consulting 3Szanowni Pastwo','SPIS, TABEL','Gmina Gubczyce pooona w poudniowo-zachodniej czci Polski na terenie wojewdztwa',
            '3Strategia Rozwoju Gminy Byczyna na lata 2015-2020','3ZESP DS. OPRACOWANIA STRATEGII ROZWOJU GMINY KORFANTW',
            '"Zabytkowy koci pw. w. Anny, oleskie maso i mietankwka,','1Strategia Rozwoju Gminy Tuowice na lata: 2022-2030',
            'Autor fotografii na okadce: Damian Wojtowicz','2Strategia Rozwoju Gminy Koaczyce do 2030 roku',
            'Strategia Rozwoju Gminy Przecaw na lata 2016-2026 jest opracowaniem, ktrego tre','JKU Konsulting Sp. z o.o. 6Strategia Zrwnowaonego Rozwoju Gminy i Miasta Nisko na lata 2014-2020',
            '3Strategia Rozwoju 2021-2030','3Strategia Rozwoju Gminy Gogw','3Syntetyczne podsumowanie dokumentu SRMT 2021-2027',
            'Gmina apy obejmuje powierzchni','3Plan Rozwoju Lokalnego Gminy Szczuczyn na lata 2008 - 2015','Strategia Rozwoju Gminy Jedwabne na lata 20212030',
            '2PLAN ROZWOJU LOKALNEGO GMINY NOWOGRD NA LATA 2016 2022','Skuteczno w wykorzystywaniu funduszy strukturalnych Unii Europejskiej',
            'Niniejszy Raport o Stanie Gminy stanowi pierwszy etap prac nad Strategi','Proces prac nad Strategi Rozwoju Gminy Ciechanowiec na lata 2021 - 2030',
            'Gwnym celem opracowania jest nakrelenie kierunkw i priorytetw rozwoju gminy Czyew na lata 2009-2017','Wykaz stosowanych skrtw',
            '2Stan gminy zosta oceniony wedug dostpnych, najnowszych danych. Podstaw do analiz','Strategia Rozwoju Gminy Nowy Dwr Gdaski na lata 2021-2030',
            'Podjcie prac nad Strategi Zrwnowaonego Rozwoju Miasta Jastarni','Strategia rozwoju gminy jest zestawieniem celw, potrzeb i sposobw ich',
            'Id: . Uchwalony Strona 4WSTP','Sopot z Gdaskiem i Gdyni tworz rdze Obszaru Metropolitalnego Gdask-Gdynia-Sopot',
            'Strategia Rozwoju Gminy Sonicowice na lata 2013  2025','Obserwacja procesw spoecznych, gospodarczych i kulturowych, ktrych Samorzd',
            '3Strategia Rozwizywania Problemw Spoecznych Miasta Imielin','Dokument Strategia Rozwoju Chorzowa do 2030 roku stanowi rezultat czci projektu pod nazw',
            '4Przedkadana Pastwu Strategia Zintegrowanego i Zrwnowaonego Rozwoju Jaworzna','Strona 4 z 157STRATEGIA ROZWOJU GMINY WODZISAW NA LATA 2015-2025',
            'Zarzdzanie gmin wie si z podejmowaniem wielu dziaa i decyzji wanych dla','LISTA DOKUMENTW GMINY MORAWICA O ZNACZENIU STRATEGICZNYM',
            'STRATEGIA ROZWOJU GMINY IWANISKA NA LATA 2017-2025','Id: . Uchwalony Strona 4Strategi a Rozwoju Gminy Poaniec na lata 2021 - 2030 ',
            'Z wielk przyjemnoci przedstawiam Pastwu Strategi Rozwoju Miasta Braniewa do','Strategia Zrwnowaonego Rozwoju Miasta i Gminy Pienino 31. Proces tworzenia strategii i jego uczestnicy.',
            'Luty 2016r. 2Program Rozwoju Miasta i Gminy Tolkmicko na lata 2016-2025','Miasto Ek i Gmina Ek wsppracuj w ramach Miejskiego Obszaru Funkcjonalnego Eku',
            'Miasta i Gminy Ryn na lata 20212031','Zarzdzanie Gmin wie si z podejmowaniem szeregu dziaa i decyzji wanych dla',
            'la V 2. V a Va Va Va WAGA WW. WE WZA U WAGA CU','2Strategia rozwoju spoeczno-gospodarczego gminy Mrgowo do 2025 r.  raport z realizacji',
            'Strona 2Strategia Rozwoju Gminy Mikoajki do roku 2025  PROJEKT','EKOTOP dr in. Roman Sobczyk',
            'Dotychczas obowizujcym dokumentem strategicznym w Gminie Czerniejewo bya uchwalona','2Strategia Rozwoju Gminy Gosty na lata 2022-2030',
            'Gmina erkw ley w wojewdztwie wielkopolskim w powiecie jarociskim i ssiaduje','W zastosowanym ujciu proces planowania strategicznego obejmuje sze nastpujcych',
            '3Strategia Rozwoju Gminy Zduny na lata 2016-2023','4PLAN ROZWOJU LOKALNEGO GMINY RYDZYNA NA LATA 2016-2022',
            '3Mosina 2020+ [STRATEGIA ROZWOJU GMINY MOSINA]', 'Strategia Rozwoju Miast i Gminy Ostrorg to dokument okrelajcy strategi dziaa',
            'Prezentowane w diagnozie dane dotycz obszaru gminy miejsko-wiejskiej Wrzenia','Strategia rozwoju Konina powstaje jako aktualizacja ju istniejcego dokumentu z 2007',
            '3 | St ro naStrategia Rozwoju Gminy Choszczno na lata 2021-2030', 'Plan Rozwoju Lokalnego Gminy Drawno to dokument zawierajcy najwaniejsze',
            'Strategia Rozwoju Gminy Czaplinek na lata 2014  2020I. Raport o stanie Gminy Czaplinek','2 | St ro naStrategia Rozwoju Gminy Zocieniec na lata 2023-2028',
            '2Strategia Goleniw 2014-2023','Strategia Rozwoju Gminy Maszewo do 2030 roku','Spis map, rysunkw, tabel i wykresw',
            '2Strategia/ Program Rozwoju Gminy Chojna','2Strategia Rozwoju Spoeczno  Gospodarczego Gminy Mieszkowice na lata 2014 - 2024',
            ',,Strategia Rozwoju Gminy Dziwnw na lata 2016- 2025 jest dokumentem okrelajcym"','4Strategia Rozwoju Gminy Kamie Pomorski na lata 2023-2030',
            'Przyszo ma wiele nazw. Dla sabych oznacza nieuchronno, dla bojaliwych nieznane','3Strategia Rozwoju Gminy Mielno na lata 2014-2022',
            'Id: . PodpisanyStrategia Rozwoju Gminy Dbno na lata 2023-2030','XVIII.	Zadania do realizacji w latach 2007-2020','Podstaw rozwoju gminy Lipiany jest strategia, ktra okrela, misj, cele i kierunki dziaania',
            '2Strategia rozwoju gminy Pyrzyce na lata 2015-2025','56GWNE WNIOSKI Z DIAGNOZY','Strategia rozwoju Gminy Biay Br na lata 2019 -2025 jest elementem polityki rozwoju',
            'Indeks rysunkw',' Analiza otoczenia oraz stanu gminy i miasta Mirosawiec','PROGRAM 1 (PR-1) ZRWNOWAONY ROZWJ FUNKCJI',
            'W 2008 roku Rada Miasta Szczecina (RM) podja uchwa1 o aktualizacji Strategii Rozwoju Szczecina, przyjtej',
            'Strategia Rozwoju Gminy Dziwnw na lata 2016- 2025 jest dokumentem okrelajcym','Planowanie i zarz dzanie strategiczne rozwoju jednostek terytorialnych jest',
            'Strategia rozwoju gminy jest drugim obok studium uwarunkowa i kierunkw zagospodarowania przestrzennego gminy dokumentem wyznaczajcym kierunki rozwoju obszaru',
            'Zabytkowy koci pw. w. Anny, oleskie maso i mietankwka, Rynek, stadion i Promenada oraz','Projekt: Nowoczesne Zarzdzanie Publiczne o numerze POKL.05.02.01-00-025/12, realizowany w ramach Programu',
            '4Strategia rozwoju Gminy Rzepin'
        ]
        extracted_text = extract_section(text, start_keywords, end_keywords)
        if extracted_text:
            keywords = ['cel', 'celw', 'wizja', 'misja']
            processed_text_with_pages = process_text_with_pages(extracted_text, keywords)
            results = []
            for sentence, start_page, end_page in processed_text_with_pages:
                if pdf:
                    extracted_text = extract_pages(pdf, [(start_page, end_page)])
                else:
                    extracted_text = ''
                results.append({
                    'file_name': os.path.basename(file_path),
                    'sentence': sentence,
                    'start_page': start_page,
                    'end_page': end_page,
                    'extracted_text': extracted_text
                })
            return results
    return None

def main():
    file_path = 'D:/AA Praktyki/strategy/1020043.pdf' 
    results = process_single_file(file_path)
    
    if results:
        for result in results:
            print(f"Plik: {result['file_name']}")
            print(f"Nagwek: {result['sentence']}")
            print(f"Strony: {result['start_page']}-{result['end_page']}")
            print(result['extracted_text'])
            print("\n" + "="*80 + "\n")
        
        df = pd.DataFrame(results)
        df.to_csv('extracted_texts.csv', index=False)
        print(f"Wyniki zapisane do pliku extracted_texts.csv")
    else:
        print("Nie znaleziono adnych wynikw.")

if __name__ == "__main__":
    main()
```

# Przypisanie brany do danego fragmentu tekstu 

```python
import pandas as pd
import pytesseract
import spacy
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import re

nlp = spacy.load("pl_core_news_lg")

def remove_hyphenation(text):
    if isinstance(text, str): 
        text = re.sub(r'(\w+)-\n(\w+)', r'\1\2', text)
        text = text.replace('\n', ' ')
    else:
        text = '' 
    return text

def clean_text(text):
    if isinstance(text, str): 
        text = re.sub(r'[^\w\s]', '', text)
        text = re.sub(r'\d+', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
    else:
        text = ''  
    return text

def lemmatize_text(text):
    if isinstance(text, str): 
        try:
            doc = nlp(text)
            lemmas = [token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct]
            return ' '.join(lemmas), lemmas
        except Exception:
            return "", []
    return "", []

csv_path = 'C:/Users/angel/extracted_texts.csv'  
df_extracted = pd.read_csv(csv_path)
xlsx_path = 'D:/AA Praktyki/excel/brana_nazwa_rozsz.xlsx'
df = pd.read_excel(xlsx_path, engine='openpyxl')

def lemmatize_nazwa_column(df):
    df['nazwa_lematyzowana'], df['lemmas'] = zip(*df['nazwa'].apply(lambda x: lemmatize_text(x) if isinstance(x, str) else (x, [])))
    return df

df = lemmatize_nazwa_column(df)
all_results = []
extracted_words = []
lemmatized_words = []

for index, row in df_extracted.iterrows():
    file_name = row['file_name']
    extracted_text = row['extracted_text']
    
    cleaned_text = remove_hyphenation(extracted_text)
    cleaned_text = clean_text(cleaned_text)
    lemmatized_text, lemmas = lemmatize_text(cleaned_text)
    
    if lemmatized_text:
        vectorizer = TfidfVectorizer()
        tfidf_matrix = vectorizer.fit_transform([lemmatized_text] + df['nazwa_lematyzowana'].tolist())
        similarity_matrix = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])
        temp_df = pd.DataFrame({'file_name': file_name, 'brana': df['brana'], 'match_score': similarity_matrix[0]})
        data = temp_df[['match_score']].values
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(data)
        dbscan = DBSCAN(eps=0.5, min_samples=2)
        clusters = dbscan.fit_predict(data_scaled)
        temp_df['cluster'] = clusters
        cluster_mean_scores = temp_df.groupby(['file_name', 'cluster'])['match_score'].mean().reset_index()
        cluster_mean_scores.columns = ['file_name', 'cluster', 'mean_match_score']
        min_mean_scores = cluster_mean_scores.groupby('file_name')['mean_match_score'].transform('min')
        cluster_mean_scores['is_min'] = cluster_mean_scores['mean_match_score'] == min_mean_scores
        filtered_clusters = cluster_mean_scores[~cluster_mean_scores['is_min']]
        final_results = temp_df.merge(filtered_clusters[['file_name', 'cluster']], on=['file_name', 'cluster'], how='inner')
        all_results.append(final_results)
        extracted_words.extend([(file_name, word) for word in cleaned_text.split()])
        lemmatized_words.extend([(file_name, lemma) for lemma in lemmas])
if all_results: 
    all_results_df = pd.concat(all_results, ignore_index=True)
    extracted_words_df = pd.DataFrame(extracted_words, columns=['file_name', 'word'])
    lemmatized_words_df = pd.DataFrame(lemmatized_words, columns=['file_name', 'lemma'])
    extracted_words_df.to_csv('extracted_words.csv', index=False)
    lemmatized_words_df.to_csv('lemmatized_words.csv', index=False)
    branche_count = all_results_df.groupby('file_name')['brana'].nunique().reset_index(name='branche_count')

    plt.figure(figsize=(10, 6))
    plt.hist(branche_count['branche_count'], bins=20, edgecolor='k', alpha=0.7)
    plt.xlabel('Liczba bran')
    plt.ylabel('Liczba gmin')
    plt.title('Histogram liczby bran w gminach po filtracji')
    plt.grid(axis='y', alpha=0.75)
    plt.show()

    all_results_df.to_csv('filtered_clusters_results_all_files.csv', index=False)
else:
    print("Brak wynikw do przetworzenia.")

```
