```python
import os
import re
import pandas as pd
import pdfplumber
from docx import Document
import textract

def read_pdf(file_path):
    try:
        with pdfplumber.open(file_path) as pdf:
            text = ''
            for page in pdf.pages:
                text += page.extract_text()
        return text
    except Exception as e:
        print(f"Nie udało się odczytać pliku PDF: {file_path}. Błąd: {e}")
        return ""

def read_txt(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
        return text
    except Exception as e:
        print(f"Nie udało się odczytać pliku TXT: {file_path}. Błąd: {e}")
        return ""

def read_docx(file_path):
    try:
        doc = Document(file_path)
        text = ''
        for paragraph in doc.paragraphs:
            text += paragraph.text + '\n'
        return text
    except Exception as e:
        print(f"Nie udało się odczytać pliku DOCX: {file_path}. Błąd: {e}")
        return ""

def read_doc(file_path):
    try:
        text = textract.process(file_path).decode('utf-8')
        return text
    except Exception as e:
        print(f"Nie udało się odczytać pliku DOC: {file_path}. Błąd: {e}")
        return ""

def combine_hyphenated_words(text):
    return re.sub(r'(\w+)-\n(\w+)', r'\1\2', text)

def clean_text(text):
    cleaned_lines = []
    skip_section = False
    for line in text.split('\n'):
        if re.search(r'\bid:.*?podpisany\b', line, flags=re.IGNORECASE):
            skip_section = True
        elif line.strip() == '':
            skip_section = False
        if not skip_section and not re.match(r'^[\s\S]*—{10,}[\s\S]*$', line):
            cleaned_lines.append(line)
    return '\n'.join(cleaned_lines)

def extract_section(text, start_keywords, end_keywords):
    start_index = -1
    for keyword in start_keywords:
        index = text.lower().find(keyword.lower())
        if index != -1:
            start_index = index
            break
    
    if start_index == -1:
        return None

    end_index = len(text)
    for end_keyword in end_keywords:
        idx = text.lower().find(end_keyword.lower(), start_index)
        if idx != -1 and idx < end_index:
            end_index = idx

    if start_index != -1 and end_index != -1:
        extracted_text = text[start_index:end_index]
        extracted_text = clean_text(extracted_text)
        extracted_text = re.sub(r'\b[0-9a-fA-F]{8}-[0-9a-fA-F]{4}-[0-9a-fA-F]{4}-[0-9a-fA-F]{4}-[0-9a-fA-F]{12}\b', '', extracted_text)
        return extracted_text
    return None

def find_sentences_with_keywords(text, keywords):
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', text)
    matching_sentences = [sentence for sentence in sentences if any(keyword in sentence.lower() for keyword in keywords)]
    return matching_sentences

def extract_page_numbers(text):
    end_page_numbers = re.findall(r'\s*(\d+)\s*\[strona\s*(\d+)\]', text, re.IGNORECASE)
    start_page_numbers = re.findall(r'^\s*(\d+)\s*$', text, re.MULTILINE)
    
    formatted_end_pages = []
    for match in end_page_numbers:
        prev_page, curr_page = map(int, match)
        formatted_end_pages.append(f"[Strona {prev_page}-{curr_page}]")
    
    return start_page_numbers, formatted_end_pages

def process_text_with_pages(text):
    sentences = re.split(r'\n+', text)
    start_page_numbers, end_page_numbers = extract_page_numbers(text)
    
    combined_results = []
    page_index = 0
    
    for sentence in sentences:
        if page_index < len(start_page_numbers):
            combined_results.append(f"{sentence.strip()} [Strona {start_page_numbers[page_index]}]")
            page_index += 1
        else:
            if page_index < len(end_page_numbers):
                combined_results.append(f"{sentence.strip()} {end_page_numbers[page_index]}")
                page_index += 1
            else:
                combined_results.append(sentence.strip())

    return combined_results

def process_file(file_path):
    if file_path.endswith('.pdf'):
        text = read_pdf(file_path)
    elif file_path.endswith('.txt'):
        text = read_txt(file_path)
    elif file_path.endswith('.docx'):
        text = read_docx(file_path)
    elif file_path.endswith('.doc'):
        text = read_doc(file_path)
    else:
        print(f"Nieobsługiwany format pliku: {file_path}")
        return None

    if text:
        text = combine_hyphenated_words(text)

        start_keywords = ['spis treści', 'zawartość', 'spis zawartości','Spis teści' , 'SPIS TEŚCI']
        end_keywords = [
            'załączniki', 'załącznik', 'spis tabel','Spis tabel', 'spis obrazów', 'spis map',
            'strategia rozwoju gminy i miasta', 'Spis wykresów','strategia rozwoju miasta i gminy',
            'spis map i tabel', 'spis rysunków i tabel', 'strategia rozwoju miasta',
            'spis ilustracji', 'wykaz tabel', 'opis konsultacji społecznych',
            'perspektywą do roku 2030', 'system wdrażania strategii rozwoju gminy wieruszów',
            'wieloletni plan inwestycyjny', 'srotegorotwju miśta gminy tonowo pomorskie olet',
            '30strategia rozwoju gminy i miasta gryfów śląski', 'wdrażanie strategii',
            'spistabei rysunków! wykresów', 'spójność z krajowymi i regionalnymi dokumentami strategicznymi',
            'system wdrażania, monitorowania i ramy finansowe','emce re śnie owe w. zbiór nowak com (panom sa. za)',
            '3Strategia Rozwoju Gminy Bolesławiec na lata 2023- 2028','Spis map, tabel, wykresów i zdjęć',
            'Spis map, tabel, wykresów i zdjęć','5Program Rozwoju Gminy Miejskiej Kowary na lata 2021-2030',
            '1STRATEGIA ROZWOJU GMINY KUDOWA-ZDRÓJ NA LATA 2021-2027','Strategia rozwoju Gminy Miejskiej Nowa Ruda na lata 2014-2020',
            'Strategii Rozwoju Gminy Polanica-Zdrój','Spis map, tabel, wykresów i zdjąć UM A','ZESPÓŁ DS. OPRACOWANIA STRATEGII ROZWOJU MIASTA ŚWIEBODZICE',
            'Zespół ds. opracowania Strategii rozwoju Gminy Głuszyca','4 5PRZEDSIĘBIORCZY WROCŁAW','SPIS RYSUNKÓW','Strategia rozwoju jest dokumentem planistycznych, o charakterze długofalowym, określającym kierunki i',
            'Regionalne Centrum Przedsiębiorczości','2Strategia Rozwoju Gminy Janikowo','2Strategia Rozwoju Gminy Janikowo na lata 2014-2020',
            'mgr Adam Stańczyk','2Czas na Nowe! Strategia Rozwoju Gminy Nowe na lata 2012-2022','Oddajemy w Państwa ręce Strategię Rozwoju Gminy Świecie do 2030 roku.',
            'Założenia rozwoju powiatu tucholskiego','OBSZAR I CZAS REALIZACJI PLANU ROZWOJU LOKALNEGO','Chcę podkreślić, że Strategia Rozwoju Bydgoszczy do 2030 roku','Spis wykresów','Miasto Rejowiec Fabryczny',
            '- 1 -Strategia Rozwoju Gminy Siedliszcze','Strategia Rozwoju Gminy Modliborzyce','"Strategia Rozwoju Gminy Modliborzyce na lata 2020-2025','Strategia Rozwoju Gminy Modliborzyce na lata 2020-2025 w postaci dokumentu, to określenie',
            '"Spis wykresów, tabel, rysunków','EKO-GEO Consulting Marek Biłyk','3Strategia rozwoju Gminy Poniatowa na lata 2015-2025',
            '2Strategia rozwoju Gminy Łaszczów na lata 2007 – 2020','Spójność Strategii Lublin 2030 z dokumentami wyższego rzędu','Agencja Rozwoju Regionalnego S.A. w Gorzowie Wielkopolskim',
            'Zawartość programu interwencji w ramach','Zarządzanie strategiczne jednostką samorządu terytorialnego jest','Niniejszy dokument został stworzony w celu przedstawienia kierunków rozwoju Gminy Ośno Lubuskie',
            'Stron a 2Strategia rozwoju gminy Świebodzin 2030','Strategia rozwoju gminy Sulechów 2030 (zwana dalej strategią) stanowi kluczowy dokument',
            'Strategia rozwoju gminy Szprotawa na lata 2015- 2023 jest najważniejszym dokumentem','15. Spis rysunków, tabel i wykresów','Strategia Rozwoju Gminy to dokument, który w jasny sposób wskazuje kierunki',
            'Głównym motywem opracowania Strategii Rozwoju gminy Rzgów była potrzeba stworzenia','Strategia Rozwoju Powiatu Poddębickiego 2020 jest','2MMIIAASSTTOO II GGMMIINNAA PPRRZZEEDDBBÓÓRRZZ',
            '4STRATEGIA ROZWOJU GMINY WIELUŃ 2025','1Strategia Rozwoju Gminy Miasta Bochnia na lata 2022-2030','Strategia Rozwoju Gminy Alwernia na lata 2022-2030, jest dokumentem, który określa',
            '3Strategia Rozwoju Gminy Trzebinia na lata 2014-2020','4STRATEGIA ROZWOJU GMINY SZCZUCIN 2014 - 2020',
            'Strategia Rozwoju Gminy Świątniki Górne na lata 2015-2020+ 2Wprowadzenie','2Strategia Rozwoju Gminy Czarny Dunajec na lata 2021-2030',
            '3Strategia Rozwoju Gminy Wolbrom','System wdrażania Strategii Rozwoju Gminy Zator na lata 2014-2022',
            'STRATEGIA ROZWOJU GMINY KOSZYCE NA LATA 20212026','GMINY NOWE BRZESKOSTRATEGIA ROZWOJU GMINY NOWE BRZESKO',
            'Rysunek','Strategia Rozwoju Gminy Radłów na lata 2021-2030 1Wprowadzenie','"Projekt: „Nowoczesne Zarządzanie Publiczne” o numerze POKL.05.02.01-00-025/12, realizowany w ramach Programu"',
            '3C I. W','Aktualne dokumenty SZJ znajdują się w sieci komputerowej w portalu pod adresem http://qsystem.nowysacz.pl Dokumenty w wersji 3',
            'Strona 4 z 99Strategia Rozwoju Gminy Wyśmierzyce na lata 2021-2030','Strona 5STRATEGIA ROZWOJU GMINY ŻELECHÓW NA LATA 2014 – 2020',
            'Strategia Rozwoju Gminy Miasta Gostynina jest dokumentem zawierającym długookresowy','Gminy Warka na lata 2005-2013 – jest sformułowanie syntetycznego dokumentu',
            'Strategia Rozwoju Gminy Kozienice na lata 2018-2030jest kierunkowo-programowym planem','Październik 2007 2Zaktualizowana Strategia Rozwoju Gminy Miejskiej Legionowo do 2015 roku',
            '2Strategia Rozwoju Gminy Solec nad Wisłą na lata 2007 - 2020','SPIS MAP','Strategia Zrównoważonego Rozwoju Gminy Halinów na lata 2016-2025',
            'Plan Rozwoju Lokalnego to dokument, który przedstaw ia zintegrowane i kompleksowe dzia łania','STRATEGIA ZRÓWNOWAŻONEGO ROZWOJU MIASTA I GMINY ZAKROCZYM DO 2020 ROKU',
            '3STRATEGIA ROZWOJU SPOŁECZNO-GOSPODARCZEGO MIASTA OTWOCKA','3STRATEGIA ROZWOJU GMINY SOCHOCIN NA LATA 2016-2020 z perspektywą do roku 2025',
            '3Strategia ZrównowaŜonego Rozwoju Miasta Piastowa do 2020 roku','Strona | 1STRATEGIA ROZWOJU GMINY MIASTO PIONKI NA LATA 2016-2022',
            'Spis tablic','W myśl rozwiązań ustawowych Władze Gminy Łomianki! Zobowiązane są do działań na','E O2OCHAOC 4724477 UE FTSWORZKL8. Podpeay son',
            'Przesłanki aktualizacji Strategii Rozwoju Gminy Zwoleń','Rozwój regionalny i lokalny ma szczególne znaczenie dla procesów wzrostu gospodarczego',
            'Opracowanie: JODKA Consulting 3Szanowni Państwo','SPIS, TABEL','Gmina Głubczyce położona w południowo-zachodniej części Polski na terenie województwa',
            '3Strategia Rozwoju Gminy Byczyna na lata 2015-2020','3ZESPÓŁ DS. OPRACOWANIA STRATEGII ROZWOJU GMINY KORFANTÓW',
            '"Zabytkowy kościół pw. św. Anny, oleskie masło i śmietankówka,','1Strategia Rozwoju Gminy Tułowice na lata: 2022-2030',
            'Autor fotografii na okładce: Damian Wojtowicz','2Strategia Rozwoju Gminy Kołaczyce do 2030 roku',
            'Strategia Rozwoju Gminy Przecław na lata 2016-2026 jest opracowaniem, którego treść','JKU Konsulting Sp. z o.o. 6„Strategia Zrównoważonego Rozwoju Gminy i Miasta Nisko na lata 2014-2020',
            '3Strategia Rozwoju 2021-2030','3Strategia Rozwoju Gminy Głogów','3Syntetyczne podsumowanie dokumentu SRMT 2021-2027',
            'Gmina Łapy obejmuje powierzchnię','3Plan Rozwoju Lokalnego Gminy Szczuczyn na lata 2008 - 2015','Strategia Rozwoju Gminy Jedwabne na lata 2021–2030',
            '2PLAN ROZWOJU LOKALNEGO GMINY NOWOGRÓD NA LATA 2016– 2022','Skuteczność w wykorzystywaniu funduszy strukturalnych Unii Europejskiej',
            'Niniejszy „Raport o Stanie Gminy” stanowi pierwszy etap prac nad Strategią','Proces prac nad Strategią Rozwoju Gminy Ciechanowiec na lata 2021 - 2030',
            'Głównym celem opracowania jest nakreślenie kierunków i priorytetów rozwoju gminy Czyżew na lata 2009-2017','Wykaz stosowanych skrótów',
            '2Stan gminy został oceniony według dostępnych, najnowszych danych. Podstawę do analiz','Strategia Rozwoju Gminy Nowy Dwór Gdański na lata 2021-2030',
            'Podjęcie prac nad Strategią ZrównowaŜonego Rozwoju Miasta Jastarni','Strategia rozwoju gminy jest zestawieniem celów, potrzeb i sposobów ich',
            'Id: . Uchwalony Strona 4WSTĘP','Sopot z Gdańskiem i Gdynią tworzą rdzeń Obszaru Metropolitalnego Gdańsk-Gdynia-Sopot',
            'Strategia Rozwoju Gminy Sośnicowice na lata 2013 – 2025','Obserwacja procesów społecznych, gospodarczych i kulturowych, których Samorząd',
            '3Strategia Rozwiązywania Problemów Społecznych Miasta Imielin','Dokument „Strategia Rozwoju Chorzowa do 2030 roku” stanowi rezultat części projektu pod nazwą',
            '4Przedkładana Państwu Strategia Zintegrowanego i Zrównoważonego Rozwoju Jaworzna','Strona 4 z 157STRATEGIA ROZWOJU GMINY WODZISŁAW NA LATA 2015-2025',
            'Zarządzanie gminą wiąże się z podejmowaniem wielu działań i decyzji ważnych dla','LISTA DOKUMENTÓW GMINY MORAWICA O ZNACZENIU STRATEGICZNYM',
            'STRATEGIA ROZWOJU GMINY IWANISKA NA LATA 2017-2025','Id: . Uchwalony Strona 4Strategi a Rozwoju Gminy Połaniec na lata 2021 - 2030 ',
            'Z wielką przyjemnością przedstawiam Państwu „Strategię Rozwoju Miasta Braniewa do','Strategia Zrównoważonego Rozwoju Miasta i Gminy Pieniężno 31. Proces tworzenia strategii i jego uczestnicy.',
            'Luty 2016r. 2Program Rozwoju Miasta i Gminy Tolkmicko na lata 2016-2025','Miasto Ełk i Gmina Ełk współpracują w ramach Miejskiego Obszaru Funkcjonalnego Ełku',
            'Miasta i Gminy Ryn na lata 2021–2031','Zarządzanie Gminą wiąże się z podejmowaniem szeregu działań i decyzji ważnych dla',
            'la V 2. V a Va Va Va WAGA WW. WE WZA U WAGA CU','2Strategia rozwoju społeczno-gospodarczego gminy Mrągowo do 2025 r. – raport z realizacji',
            'Strona 2Strategia Rozwoju Gminy Mikołajki do roku 2025 – PROJEKT','EKOTOP dr inż. Roman Sobczyk',
            'Dotychczas obowiązującym dokumentem strategicznym w Gminie Czerniejewo była uchwalona','2Strategia Rozwoju Gminy Gostyń na lata 2022-2030',
            'Gmina Żerków leży w województwie wielkopolskim w powiecie jarocińskim i sąsiaduje','W zastosowanym ujęciu proces planowania strategicznego obejmuje sześć następujących',
            '3Strategia Rozwoju Gminy Zduny na lata 2016-2023','4PLAN ROZWOJU LOKALNEGO GMINY RYDZYNA NA LATA 2016-2022',
            '3Mosina 2020+ [STRATEGIA ROZWOJU GMINY MOSINA]', 'Strategia Rozwoju Miast i Gminy Ostroróg to dokument określający strategię działań',
            'Prezentowane w diagnozie dane dotyczą obszaru gminy miejsko-wiejskiej Września','Strategia rozwoju Konina powstaje jako aktualizacja już istniejącego dokumentu z 2007',
            '3 | St ro naStrategia Rozwoju Gminy Choszczno na lata 2021-2030', 'Plan Rozwoju Lokalnego Gminy Drawno to dokument zawierający najważniejsze',
            'Strategia Rozwoju Gminy Czaplinek na lata 2014 ‐ 2020I. Raport o stanie Gminy Czaplinek','2 | St ro naStrategia Rozwoju Gminy Złocieniec na lata 2023-2028',
            '2Strategia Goleniów 2014-2023','Strategia Rozwoju Gminy Maszewo do 2030 roku','Spis map, rysunków, tabel i wykresów',
            '2Strategia/ Program Rozwoju Gminy Chojna','2Strategia Rozwoju Społeczno – Gospodarczego Gminy Mieszkowice na lata 2014 - 2024',
            ',,Strategia Rozwoju Gminy Dziwnów na lata 2016- 2025” jest dokumentem określającym"','4Strategia Rozwoju Gminy Kamień Pomorski na lata 2023-2030',
            'Przyszłość ma wiele nazw. Dla słabych oznacza nieuchronność, dla bojaźliwych nieznane','3Strategia Rozwoju Gminy Mielno na lata 2014-2022',
            'Id: . PodpisanyStrategia Rozwoju Gminy Dębno na lata 2023-2030','XVIII.	Zadania do realizacji w latach 2007-2020','Podstawą rozwoju gminy Lipiany jest strategia, która określa, misję, cele i kierunki działania',
            '2Strategia rozwoju gminy Pyrzyce na lata 2015-2025','56GŁÓWNE WNIOSKI Z DIAGNOZY','Strategia rozwoju Gminy Biały Bór na lata 2019 -2025 jest elementem polityki rozwoju',
            'Indeks rysunków',' Analiza otoczenia oraz stanu gminy i miasta Mirosławiec','PROGRAM 1 (PR-1) ZRÓWNOWAŻONY ROZWÓJ FUNKCJI',
            'W 2008 roku Rada Miasta Szczecina (RM) podjęła uchwałę1 o aktualizacji Strategii Rozwoju Szczecina, przyjętej',
            'Strategia Rozwoju Gminy Dziwnów na lata 2016- 2025” jest dokumentem określającym','Planowanie i zarz ądzanie strategiczne rozwoju jednostek terytorialnych jest',
            'Strategia rozwoju gminy jest drugim obok studium uwarunkowań i kierunków zagospodarowania przestrzennego gminy dokumentem wyznaczającym kierunki rozwoju obszaru',
            'Zabytkowy kościół pw. św. Anny, oleskie masło i śmietankówka, Rynek, stadion i Promenada oraz','Projekt: „Nowoczesne Zarządzanie Publiczne” o numerze POKL.05.02.01-00-025/12, realizowany w ramach Programu',
            '4Strategia rozwoju Gminy Rzepin'
        ]
        extracted_text = extract_section(text, start_keywords, end_keywords)
        if extracted_text:
            keywords = ['cel', 'celów', 'wizja', 'misja']
            sentences = find_sentences_with_keywords(extracted_text, keywords)
            
            processed_text_with_pages = process_text_with_pages(extracted_text)
            
            return processed_text_with_pages
    return None

def process_directory(directory_path):
    results = []
    for root, dirs, files in os.walk(directory_path):
        for file in files:
            file_path = os.path.join(root, file)
            combined_sentences = process_file(file_path)
            if combined_sentences:
                results.append({'file_name': file, 'sentences': combined_sentences})
    return results

def save_results_to_csv(results, csv_filename):
    all_data = []
    for result in results:
        for sentence in result['sentences']:
            all_data.append({'file_name': result['file_name'], 'sentence': sentence})
    
    df = pd.DataFrame(all_data)
    df.to_csv(csv_filename, index=False, encoding='utf-8-sig')

def save_results_to_docx(results, docx_filename):
    doc = Document()
    for result in results:
        doc.add_heading(result['file_name'], level=1)
        for sentence in result['sentences']:
            doc.add_paragraph(sentence)
        doc.add_page_break()
    
    doc.save(docx_filename)

def main():
    directory_path = 'D:/AA Praktyki/strategy' 
    results = process_directory(directory_path)
    
    if results:
        csv_filename = 'spisy.csv'
        docx_filename = 'spisy.docx'
        
        save_results_to_csv(results, csv_filename)
        save_results_to_docx(results, docx_filename)
        
        print(f"Wyniki zapisane do plików {csv_filename} i {docx_filename}")
    else:
        print("Nie znaleziono żadnych wyników do zapisania.")

if __name__ == "__main__":
    main()

```

```python
import os
import re
import pandas as pd
import pdfplumber
from docx import Document

def read_pdf(file_path):
    try:
        with pdfplumber.open(file_path) as pdf:
            text = ''
            for page in pdf.pages:
                text += page.extract_text()
        return text, pdf
    except Exception as e:
        print(f"Nie udało się odczytać pliku PDF: {file_path}. Błąd: {e}")
        return "", None

def read_txt(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
        return text
    except Exception as e:
        print(f"Nie udało się odczytać pliku TXT: {file_path}. Błąd: {e}")
        return ""

def read_docx(file_path):
    try:
        doc = Document(file_path)
        text = ''
        for paragraph in doc.paragraphs:
            text += paragraph.text + '\n'
        return text
    except Exception as e:
        print(f"Nie udało się odczytać pliku DOCX: {file_path}. Błąd: {e}")
        return ""

def combine_hyphenated_words(text):
    # Combine words split by hyphens at line ends
    return re.sub(r'(\w+)-\n(\w+)', r'\1\2', text)

def clean_text(text):
    cleaned_lines = []
    skip_section = False
    for line in text.split('\n'):
        if re.search(r'\bid:.*?podpisany\b', line, flags=re.IGNORECASE):
            skip_section = True
        elif line.strip() == '':
            skip_section = False
        if not skip_section and not re.match(r'^[\s\S]*—{10,}[\s\S]*$', line):
            cleaned_lines.append(line)
    return '\n'.join(cleaned_lines)

def extract_section(text, start_keywords, end_keywords):
    start_index = -1
    for keyword in start_keywords:
        index = text.lower().find(keyword.lower())
        if index != -1:
            start_index = index
            break
    
    if start_index == -1:
        return None

    end_index = len(text)
    for end_keyword in end_keywords:
        idx = text.lower().find(end_keyword.lower(), start_index)
        if idx != -1 and idx < end_index:
            end_index = idx

    if start_index != -1 and end_index != -1:
        extracted_text = text[start_index:end_index]
        extracted_text = clean_text(extracted_text)
        extracted_text = re.sub(r'\b[0-9a-fA-F]{8}-[0-9a-fA-F]{4}-[0-9a-fA-F]{4}-[0-9a-fA-F]{4}-[0-9a-fA-F]{12}\b', '', extracted_text)
        return extracted_text
    return None

def find_sentences_with_keywords(text, keywords):
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?)\s', text)
    matching_sentences = [sentence for sentence in sentences if any(keyword in sentence.lower() for keyword in keywords)]
    return matching_sentences

def extract_page_numbers(text):
    end_page_numbers = re.findall(r'\s*(\d+)\s*$', text, re.MULTILINE)
    start_page_numbers = re.findall(r'^\s*(\d+)', text, re.MULTILINE)
    return end_page_numbers, start_page_numbers

def process_text_with_pages(text, keywords):
    sentences = re.split(r'\n+', text)
    combined_results = []
    
    for i in range(len(sentences)):
        sentence = sentences[i].strip()
        next_page = None
        
        current_page_match = re.search(r'\b(\d+)\s*$', sentence)
        if current_page_match:
            current_page = current_page_match.group(1)
        else:
            continue
        
        if i + 1 < len(sentences):
            next_sentence = sentences[i + 1].strip()
            next_page_match = re.search(r'\b(\d+)\s*$', next_sentence)
            if next_page_match:
                next_page = next_page_match.group(1)
        
        if any(keyword in sentence.lower() for keyword in keywords):
            if next_page:
                combined_results.append((sentence, int(current_page), int(next_page)))
            else:
                combined_results.append((sentence, int(current_page), int(current_page)))
    
    return combined_results

def extract_pages(pdf, page_ranges):
    extracted_text = ''
    for start_page, end_page in page_ranges:
        for page_num in range(start_page, end_page + 1):
            try:
                extracted_text += pdf.pages[page_num - 1].extract_text() + '\n'
            except IndexError:
                print(f"Strona {page_num} nie istnieje w dokumencie PDF.")
    return extracted_text

def process_file(file_path):
    if file_path.endswith('.pdf'):
        text, pdf = read_pdf(file_path)
    elif file_path.endswith('.txt'):
        text = read_txt(file_path)
        pdf = None
    elif file_path.endswith('.docx'):
        text = read_docx(file_path)
        pdf = None
    else:
        print(f"Nieobsługiwany format pliku: {file_path}")
        return None, None

    if text:
        # Combine hyphenated words first
        text = combine_hyphenated_words(text)

        start_keywords = ['spis treści', 'zawartość', 'spis zawartości']
        end_keywords = [
            'załączniki', 'załącznik', 'spis tabel','Spis tabel', 'spis obrazów', 'spis map',
            'strategia rozwoju gminy i miasta', 'Spis wykresów','strategia rozwoju miasta i gminy',
            'spis map i tabel', 'spis rysunków i tabel', 'strategia rozwoju miasta',
            'spis ilustracji', 'wykaz tabel', 'opis konsultacji społecznych',
            'perspektywą do roku 2030', 'system wdrażania strategii rozwoju gminy wieruszów',
            'wieloletni plan inwestycyjny', 'srotegorotwju miśta gminy tonowo pomorskie olet',
            '30strategia rozwoju gminy i miasta gryfów śląski', 'wdrażanie strategii',
            'spistabei rysunków! wykresów', 'spójność z krajowymi i regionalnymi dokumentami strategicznymi',
            'system wdrażania, monitorowania i ramy finansowe','emce re śnie owe w. zbiór nowak com (panom sa. za)',
            '3Strategia Rozwoju Gminy Bolesławiec na lata 2023- 2028','Spis map, tabel, wykresów i zdjęć',
            'Spis map, tabel, wykresów i zdjęć','5Program Rozwoju Gminy Miejskiej Kowary na lata 2021-2030',
            '1STRATEGIA ROZWOJU GMINY KUDOWA-ZDRÓJ NA LATA 2021-2027','Strategia rozwoju Gminy Miejskiej Nowa Ruda na lata 2014-2020',
            'Strategii Rozwoju Gminy Polanica-Zdrój','Spis map, tabel, wykresów i zdjąć UM A','ZESPÓŁ DS. OPRACOWANIA STRATEGII ROZWOJU MIASTA ŚWIEBODZICE',
            'Zespół ds. opracowania Strategii rozwoju Gminy Głuszyca','4 5PRZEDSIĘBIORCZY WROCŁAW','SPIS RYSUNKÓW','Strategia rozwoju jest dokumentem planistycznych, o charakterze długofalowym, określającym kierunki i',
            'Regionalne Centrum Przedsiębiorczości','2Strategia Rozwoju Gminy Janikowo','2Strategia Rozwoju Gminy Janikowo na lata 2014-2020',
            'mgr Adam Stańczyk','2Czas na Nowe! Strategia Rozwoju Gminy Nowe na lata 2012-2022','Oddajemy w Państwa ręce Strategię Rozwoju Gminy Świecie do 2030 roku.',
            'Założenia rozwoju powiatu tucholskiego','OBSZAR I CZAS REALIZACJI PLANU ROZWOJU LOKALNEGO','Chcę podkreślić, że Strategia Rozwoju Bydgoszczy do 2030 roku','Spis wykresów','Miasto Rejowiec Fabryczny',
            '- 1 -Strategia Rozwoju Gminy Siedliszcze','Strategia Rozwoju Gminy Modliborzyce','"Strategia Rozwoju Gminy Modliborzyce na lata 2020-2025','Strategia Rozwoju Gminy Modliborzyce na lata 2020-2025 w postaci dokumentu, to określenie',
            '"Spis wykresów, tabel, rysunków','EKO-GEO Consulting Marek Biłyk','3Strategia rozwoju Gminy Poniatowa na lata 2015-2025',
            '2Strategia rozwoju Gminy Łaszczów na lata 2007 – 2020','Spójność Strategii Lublin 2030 z dokumentami wyższego rzędu','Agencja Rozwoju Regionalnego S.A. w Gorzowie Wielkopolskim',
            'Zawartość programu interwencji w ramach','Zarządzanie strategiczne jednostką samorządu terytorialnego jest','Niniejszy dokument został stworzony w celu przedstawienia kierunków rozwoju Gminy Ośno Lubuskie',
            'Stron a 2Strategia rozwoju gminy Świebodzin 2030','Strategia rozwoju gminy Sulechów 2030 (zwana dalej strategią) stanowi kluczowy dokument',
            'Strategia rozwoju gminy Szprotawa na lata 2015- 2023 jest najważniejszym dokumentem','15. Spis rysunków, tabel i wykresów','Strategia Rozwoju Gminy to dokument, który w jasny sposób wskazuje kierunki',
            'Głównym motywem opracowania Strategii Rozwoju gminy Rzgów była potrzeba stworzenia','Strategia Rozwoju Powiatu Poddębickiego 2020 jest','2MMIIAASSTTOO II GGMMIINNAA PPRRZZEEDDBBÓÓRRZZ',
            '4STRATEGIA ROZWOJU GMINY WIELUŃ 2025','1Strategia Rozwoju Gminy Miasta Bochnia na lata 2022-2030','Strategia Rozwoju Gminy Alwernia na lata 2022-2030, jest dokumentem, który określa',
            '3Strategia Rozwoju Gminy Trzebinia na lata 2014-2020','4STRATEGIA ROZWOJU GMINY SZCZUCIN 2014 - 2020',
            'Strategia Rozwoju Gminy Świątniki Górne na lata 2015-2020+ 2Wprowadzenie','2Strategia Rozwoju Gminy Czarny Dunajec na lata 2021-2030',
            '3Strategia Rozwoju Gminy Wolbrom','System wdrażania Strategii Rozwoju Gminy Zator na lata 2014-2022',
            'STRATEGIA ROZWOJU GMINY KOSZYCE NA LATA 20212026','GMINY NOWE BRZESKOSTRATEGIA ROZWOJU GMINY NOWE BRZESKO',
            'Rysunek','Strategia Rozwoju Gminy Radłów na lata 2021-2030 1Wprowadzenie','"Projekt: „Nowoczesne Zarządzanie Publiczne” o numerze POKL.05.02.01-00-025/12, realizowany w ramach Programu"',
            '3C I. W','Aktualne dokumenty SZJ znajdują się w sieci komputerowej w portalu pod adresem http://qsystem.nowysacz.pl Dokumenty w wersji 3',
            'Strona 4 z 99Strategia Rozwoju Gminy Wyśmierzyce na lata 2021-2030','Strona 5STRATEGIA ROZWOJU GMINY ŻELECHÓW NA LATA 2014 – 2020',
            'Strategia Rozwoju Gminy Miasta Gostynina jest dokumentem zawierającym długookresowy','Gminy Warka na lata 2005-2013 – jest sformułowanie syntetycznego dokumentu',
            'Strategia Rozwoju Gminy Kozienice na lata 2018-2030jest kierunkowo-programowym planem','Październik 2007 2Zaktualizowana Strategia Rozwoju Gminy Miejskiej Legionowo do 2015 roku',
            '2Strategia Rozwoju Gminy Solec nad Wisłą na lata 2007 - 2020','SPIS MAP','Strategia Zrównoważonego Rozwoju Gminy Halinów na lata 2016-2025',
            'Plan Rozwoju Lokalnego to dokument, który przedstaw ia zintegrowane i kompleksowe dzia łania','STRATEGIA ZRÓWNOWAŻONEGO ROZWOJU MIASTA I GMINY ZAKROCZYM DO 2020 ROKU',
            '3STRATEGIA ROZWOJU SPOŁECZNO-GOSPODARCZEGO MIASTA OTWOCKA','3STRATEGIA ROZWOJU GMINY SOCHOCIN NA LATA 2016-2020 z perspektywą do roku 2025',
            '3Strategia ZrównowaŜonego Rozwoju Miasta Piastowa do 2020 roku','Strona | 1STRATEGIA ROZWOJU GMINY MIASTO PIONKI NA LATA 2016-2022',
            'Spis tablic','W myśl rozwiązań ustawowych Władze Gminy Łomianki! Zobowiązane są do działań na','E O2OCHAOC 4724477 UE FTSWORZKL8. Podpeay son',
            'Przesłanki aktualizacji Strategii Rozwoju Gminy Zwoleń','Rozwój regionalny i lokalny ma szczególne znaczenie dla procesów wzrostu gospodarczego',
            'Opracowanie: JODKA Consulting 3Szanowni Państwo','SPIS, TABEL','Gmina Głubczyce położona w południowo-zachodniej części Polski na terenie województwa',
            '3Strategia Rozwoju Gminy Byczyna na lata 2015-2020','3ZESPÓŁ DS. OPRACOWANIA STRATEGII ROZWOJU GMINY KORFANTÓW',
            '"Zabytkowy kościół pw. św. Anny, oleskie masło i śmietankówka,','1Strategia Rozwoju Gminy Tułowice na lata: 2022-2030',
            'Autor fotografii na okładce: Damian Wojtowicz','2Strategia Rozwoju Gminy Kołaczyce do 2030 roku',
            'Strategia Rozwoju Gminy Przecław na lata 2016-2026 jest opracowaniem, którego treść','JKU Konsulting Sp. z o.o. 6„Strategia Zrównoważonego Rozwoju Gminy i Miasta Nisko na lata 2014-2020',
            '3Strategia Rozwoju 2021-2030','3Strategia Rozwoju Gminy Głogów','3Syntetyczne podsumowanie dokumentu SRMT 2021-2027',
            'Gmina Łapy obejmuje powierzchnię','3Plan Rozwoju Lokalnego Gminy Szczuczyn na lata 2008 - 2015','Strategia Rozwoju Gminy Jedwabne na lata 2021–2030',
            '2PLAN ROZWOJU LOKALNEGO GMINY NOWOGRÓD NA LATA 2016– 2022','Skuteczność w wykorzystywaniu funduszy strukturalnych Unii Europejskiej',
            'Niniejszy „Raport o Stanie Gminy” stanowi pierwszy etap prac nad Strategią','Proces prac nad Strategią Rozwoju Gminy Ciechanowiec na lata 2021 - 2030',
            'Głównym celem opracowania jest nakreślenie kierunków i priorytetów rozwoju gminy Czyżew na lata 2009-2017','Wykaz stosowanych skrótów',
            '2Stan gminy został oceniony według dostępnych, najnowszych danych. Podstawę do analiz','Strategia Rozwoju Gminy Nowy Dwór Gdański na lata 2021-2030',
            'Podjęcie prac nad Strategią ZrównowaŜonego Rozwoju Miasta Jastarni','Strategia rozwoju gminy jest zestawieniem celów, potrzeb i sposobów ich',
            'Id: . Uchwalony Strona 4WSTĘP','Sopot z Gdańskiem i Gdynią tworzą rdzeń Obszaru Metropolitalnego Gdańsk-Gdynia-Sopot',
            'Strategia Rozwoju Gminy Sośnicowice na lata 2013 – 2025','Obserwacja procesów społecznych, gospodarczych i kulturowych, których Samorząd',
            '3Strategia Rozwiązywania Problemów Społecznych Miasta Imielin','Dokument „Strategia Rozwoju Chorzowa do 2030 roku” stanowi rezultat części projektu pod nazwą',
            '4Przedkładana Państwu Strategia Zintegrowanego i Zrównoważonego Rozwoju Jaworzna','Strona 4 z 157STRATEGIA ROZWOJU GMINY WODZISŁAW NA LATA 2015-2025',
            'Zarządzanie gminą wiąże się z podejmowaniem wielu działań i decyzji ważnych dla','LISTA DOKUMENTÓW GMINY MORAWICA O ZNACZENIU STRATEGICZNYM',
            'STRATEGIA ROZWOJU GMINY IWANISKA NA LATA 2017-2025','Id: . Uchwalony Strona 4Strategi a Rozwoju Gminy Połaniec na lata 2021 - 2030 ',
            'Z wielką przyjemnością przedstawiam Państwu „Strategię Rozwoju Miasta Braniewa do','Strategia Zrównoważonego Rozwoju Miasta i Gminy Pieniężno 31. Proces tworzenia strategii i jego uczestnicy.',
            'Luty 2016r. 2Program Rozwoju Miasta i Gminy Tolkmicko na lata 2016-2025','Miasto Ełk i Gmina Ełk współpracują w ramach Miejskiego Obszaru Funkcjonalnego Ełku',
            'Miasta i Gminy Ryn na lata 2021–2031','Zarządzanie Gminą wiąże się z podejmowaniem szeregu działań i decyzji ważnych dla',
            'la V 2. V a Va Va Va WAGA WW. WE WZA U WAGA CU','2Strategia rozwoju społeczno-gospodarczego gminy Mrągowo do 2025 r. – raport z realizacji',
            'Strona 2Strategia Rozwoju Gminy Mikołajki do roku 2025 – PROJEKT','EKOTOP dr inż. Roman Sobczyk',
            'Dotychczas obowiązującym dokumentem strategicznym w Gminie Czerniejewo była uchwalona','2Strategia Rozwoju Gminy Gostyń na lata 2022-2030',
            'Gmina Żerków leży w województwie wielkopolskim w powiecie jarocińskim i sąsiaduje','W zastosowanym ujęciu proces planowania strategicznego obejmuje sześć następujących',
            '3Strategia Rozwoju Gminy Zduny na lata 2016-2023','4PLAN ROZWOJU LOKALNEGO GMINY RYDZYNA NA LATA 2016-2022',
            '3Mosina 2020+ [STRATEGIA ROZWOJU GMINY MOSINA]', 'Strategia Rozwoju Miast i Gminy Ostroróg to dokument określający strategię działań',
            'Prezentowane w diagnozie dane dotyczą obszaru gminy miejsko-wiejskiej Września','Strategia rozwoju Konina powstaje jako aktualizacja już istniejącego dokumentu z 2007',
            '3 | St ro naStrategia Rozwoju Gminy Choszczno na lata 2021-2030', 'Plan Rozwoju Lokalnego Gminy Drawno to dokument zawierający najważniejsze',
            'Strategia Rozwoju Gminy Czaplinek na lata 2014 ‐ 2020I. Raport o stanie Gminy Czaplinek','2 | St ro naStrategia Rozwoju Gminy Złocieniec na lata 2023-2028',
            '2Strategia Goleniów 2014-2023','Strategia Rozwoju Gminy Maszewo do 2030 roku','Spis map, rysunków, tabel i wykresów',
            '2Strategia/ Program Rozwoju Gminy Chojna','2Strategia Rozwoju Społeczno – Gospodarczego Gminy Mieszkowice na lata 2014 - 2024',
            ',,Strategia Rozwoju Gminy Dziwnów na lata 2016- 2025” jest dokumentem określającym"','4Strategia Rozwoju Gminy Kamień Pomorski na lata 2023-2030',
            'Przyszłość ma wiele nazw. Dla słabych oznacza nieuchronność, dla bojaźliwych nieznane','3Strategia Rozwoju Gminy Mielno na lata 2014-2022',
            'Id: . PodpisanyStrategia Rozwoju Gminy Dębno na lata 2023-2030','XVIII.	Zadania do realizacji w latach 2007-2020','Podstawą rozwoju gminy Lipiany jest strategia, która określa, misję, cele i kierunki działania',
            '2Strategia rozwoju gminy Pyrzyce na lata 2015-2025','56GŁÓWNE WNIOSKI Z DIAGNOZY','Strategia rozwoju Gminy Biały Bór na lata 2019 -2025 jest elementem polityki rozwoju',
            'Indeks rysunków',' Analiza otoczenia oraz stanu gminy i miasta Mirosławiec','PROGRAM 1 (PR-1) ZRÓWNOWAŻONY ROZWÓJ FUNKCJI',
            'W 2008 roku Rada Miasta Szczecina (RM) podjęła uchwałę1 o aktualizacji Strategii Rozwoju Szczecina, przyjętej',
            'Strategia Rozwoju Gminy Dziwnów na lata 2016- 2025” jest dokumentem określającym','Planowanie i zarz ądzanie strategiczne rozwoju jednostek terytorialnych jest',
            'Strategia rozwoju gminy jest drugim obok studium uwarunkowań i kierunków zagospodarowania przestrzennego gminy dokumentem wyznaczającym kierunki rozwoju obszaru',
            'Zabytkowy kościół pw. św. Anny, oleskie masło i śmietankówka, Rynek, stadion i Promenada oraz','Projekt: „Nowoczesne Zarządzanie Publiczne” o numerze POKL.05.02.01-00-025/12, realizowany w ramach Programu',
            '4Strategia rozwoju Gminy Rzepin'
        ]
        extracted_text = extract_section(text, start_keywords, end_keywords)
        if extracted_text:
            keywords = ['cel', 'celów', 'wizja', 'misja']
            processed_text_with_pages = process_text_with_pages(extracted_text, keywords)
            return processed_text_with_pages, pdf
    return None, None

def process_directory(directory_path):
    results = []
    for root, dirs, files in os.walk(directory_path):
        for file in files:
            file_path = os.path.join(root, file)
            combined_sentences, pdf = process_file(file_path)
            if combined_sentences and pdf:
                for sentence, start_page, end_page in combined_sentences:
                    extracted_text = extract_pages(pdf, [(start_page, end_page)])
                    results.append({
                        'file_name': file,
                        'sentence': sentence,
                        'start_page': start_page,
                        'end_page': end_page,
                        'extracted_text': extracted_text
                    })
    return results

def main():
    directory_path = 'D:/AA Praktyki/strategy'  # Zmień ścieżkę na właściwą
    results = process_directory(directory_path)
    
    # Wyświetl wyniki
    for result in results:
        print(f"Plik: {result['file_name']}")
        print(f"Nagłówek: {result['sentence']}")
        print(f"Strony: {result['start_page']}-{result['end_page']}")
        print(result['extracted_text'])
        print("\n" + "="*80 + "\n")
    
    # Zapisz wyniki do pliku CSV
    df = pd.DataFrame(results)
    df.to_csv('extracted_texts.csv', index=False)
    print(f"Wyniki zapisane do pliku extracted_texts.csv")

if __name__ == "__main__":
    main()

```

```python
import os
import re
import pandas as pd
import pdfplumber
from docx import Document

def read_pdf(file_path):
    try:
        with pdfplumber.open(file_path) as pdf:
            text = ''
            for page in pdf.pages:
                text += page.extract_text()
        return text, pdf
    except Exception as e:
        print(f"Nie udało się odczytać pliku PDF: {file_path}. Błąd: {e}")
        return "", None

def read_txt(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
        return text
    except Exception as e:
        print(f"Nie udało się odczytać pliku TXT: {file_path}. Błąd: {e}")
        return ""

def read_docx(file_path):
    try:
        doc = Document(file_path)
        text = ''
        for paragraph in doc.paragraphs:
            text += paragraph.text + '\n'
        return text
    except Exception as e:
        print(f"Nie udało się odczytać pliku DOCX: {file_path}. Błąd: {e}")
        return ""

def combine_hyphenated_words(text):
    return re.sub(r'(\w+)-\n(\w+)', r'\1\2', text)

def clean_text(text):
    cleaned_lines = []
    skip_section = False
    for line in text.split('\n'):
        if re.search(r'\bid:.*?podpisany\b', line, flags=re.IGNORECASE):
            skip_section = True
        elif line.strip() == '':
            skip_section = False
        if not skip_section and not re.match(r'^[\s\S]*—{10,}[\s\S]*$', line):
            cleaned_lines.append(line)
    return '\n'.join(cleaned_lines)

def extract_section(text, start_keywords, end_keywords):
    start_index = -1
    for keyword in start_keywords:
        index = text.lower().find(keyword.lower())
        if index != -1:
            start_index = index
            break
    
    if start_index == -1:
        return None

    end_index = len(text)
    for end_keyword in end_keywords:
        idx = text.lower().find(end_keyword.lower(), start_index)
        if idx != -1 and idx < end_index:
            end_index = idx

    if start_index != -1 and end_index != -1:
        extracted_text = text[start_index:end_index]
        extracted_text = clean_text(extracted_text)
        extracted_text = re.sub(r'\b[0-9a-fA-F]{8}-[0-9a-fA-F]{4}-[0-9a-fA-F]{4}-[0-9a-fA-F]{4}-[0-9a-fA-F]{12}\b', '', extracted_text)
        return extracted_text
    return None

def process_text_with_pages(text, keywords, table_of_contents_pages=(1, 3)):
    sentences = re.split(r'\n+', text)
    combined_results = []
    num_sentences = len(sentences)

    i = 0
    while i < num_sentences:
        sentence = sentences[i].strip()
        current_page = None
        current_page_match = re.search(r'\.{3,}\s*(\d+)\s*$', sentence)
        if current_page_match:
            current_page = int(current_page_match.group(1))
            if current_page > 1000:
                i += 1
                continue  
            if table_of_contents_pages[0] <= current_page <= table_of_contents_pages[1]:
                i += 1
                continue  
        else:
            i += 1
            continue

        next_page = None
        j = i + 1
        while j < num_sentences:
            next_sentence = sentences[j].strip()
            if re.search(r'^strona \d+', next_sentence, re.IGNORECASE):
                j += 1
                continue

            next_page_match = re.search(r'\.{3,}\s*(\d+)\s*$', next_sentence)
            if next_page_match:
                next_page = int(next_page_match.group(1))
                if next_page >= current_page and next_page < 1000:
                    break
            j += 1

        if next_page:
            end_page = next_page
        else:
            end_page = current_page  
        if any(keyword in sentence.lower() for keyword in keywords):
            combined_results.append((sentence, current_page, end_page))
        i = j
    
    return combined_results

def extract_pages(pdf, page_ranges):
    extracted_text = ''
    for start_page, end_page in page_ranges:
        for page_num in range(start_page, end_page + 1):
            try:
                extracted_text += pdf.pages[page_num - 1].extract_text() + '\n'
            except IndexError:
                print(f"Strona {page_num} nie istnieje w dokumencie PDF.")
    return extracted_text

def process_single_file(file_path):
    if file_path.endswith('.pdf'):
        text, pdf = read_pdf(file_path)
    elif file_path.endswith('.txt'):
        text = read_txt(file_path)
        pdf = None
    elif file_path.endswith('.docx'):
        text = read_docx(file_path)
        pdf = None
    else:
        print(f"Nieobsługiwany format pliku: {file_path}")
        return None

    if text:
        text = combine_hyphenated_words(text)

        start_keywords = ['spis treści', 'zawartość', 'spis zawartości']
        end_keywords = [
            'załączniki', 'załącznik', 'spis tabel','Spis tabel', 'spis obrazów', 'spis map',
            'strategia rozwoju gminy i miasta', 'Spis wykresów','strategia rozwoju miasta i gminy',
            'spis map i tabel', 'spis rysunków i tabel', 'strategia rozwoju miasta',
            'spis ilustracji', 'wykaz tabel', 'opis konsultacji społecznych',
            'perspektywą do roku 2030', 'system wdrażania strategii rozwoju gminy wieruszów',
            'wieloletni plan inwestycyjny', 'srotegorotwju miśta gminy tonowo pomorskie olet',
            '30strategia rozwoju gminy i miasta gryfów śląski', 'wdrażanie strategii',
            'spistabei rysunków! wykresów', 'spójność z krajowymi i regionalnymi dokumentami strategicznymi',
            'system wdrażania, monitorowania i ramy finansowe','emce re śnie owe w. zbiór nowak com (panom sa. za)',
            '3Strategia Rozwoju Gminy Bolesławiec na lata 2023- 2028','Spis map, tabel, wykresów i zdjęć',
            'Spis map, tabel, wykresów i zdjęć','5Program Rozwoju Gminy Miejskiej Kowary na lata 2021-2030',
            '1STRATEGIA ROZWOJU GMINY KUDOWA-ZDRÓJ NA LATA 2021-2027','Strategia rozwoju Gminy Miejskiej Nowa Ruda na lata 2014-2020',
            'Strategii Rozwoju Gminy Polanica-Zdrój','Spis map, tabel, wykresów i zdjąć UM A','ZESPÓŁ DS. OPRACOWANIA STRATEGII ROZWOJU MIASTA ŚWIEBODZICE',
            'Zespół ds. opracowania Strategii rozwoju Gminy Głuszyca','4 5PRZEDSIĘBIORCZY WROCŁAW','SPIS RYSUNKÓW','Strategia rozwoju jest dokumentem planistycznych, o charakterze długofalowym, określającym kierunki i',
            'Regionalne Centrum Przedsiębiorczości','2Strategia Rozwoju Gminy Janikowo','2Strategia Rozwoju Gminy Janikowo na lata 2014-2020',
            'mgr Adam Stańczyk','2Czas na Nowe! Strategia Rozwoju Gminy Nowe na lata 2012-2022','Oddajemy w Państwa ręce Strategię Rozwoju Gminy Świecie do 2030 roku.',
            'Założenia rozwoju powiatu tucholskiego','OBSZAR I CZAS REALIZACJI PLANU ROZWOJU LOKALNEGO','Chcę podkreślić, że Strategia Rozwoju Bydgoszczy do 2030 roku','Spis wykresów','Miasto Rejowiec Fabryczny',
            '- 1 -Strategia Rozwoju Gminy Siedliszcze','Strategia Rozwoju Gminy Modliborzyce','"Strategia Rozwoju Gminy Modliborzyce na lata 2020-2025','Strategia Rozwoju Gminy Modliborzyce na lata 2020-2025 w postaci dokumentu, to określenie',
            '"Spis wykresów, tabel, rysunków','EKO-GEO Consulting Marek Biłyk','3Strategia rozwoju Gminy Poniatowa na lata 2015-2025',
            '2Strategia rozwoju Gminy Łaszczów na lata 2007 – 2020','Spójność Strategii Lublin 2030 z dokumentami wyższego rzędu','Agencja Rozwoju Regionalnego S.A. w Gorzowie Wielkopolskim',
            'Zawartość programu interwencji w ramach','Zarządzanie strategiczne jednostką samorządu terytorialnego jest','Niniejszy dokument został stworzony w celu przedstawienia kierunków rozwoju Gminy Ośno Lubuskie',
            'Stron a 2Strategia rozwoju gminy Świebodzin 2030','Strategia rozwoju gminy Sulechów 2030 (zwana dalej strategią) stanowi kluczowy dokument',
            'Strategia rozwoju gminy Szprotawa na lata 2015- 2023 jest najważniejszym dokumentem','15. Spis rysunków, tabel i wykresów','Strategia Rozwoju Gminy to dokument, który w jasny sposób wskazuje kierunki',
            'Głównym motywem opracowania Strategii Rozwoju gminy Rzgów była potrzeba stworzenia','Strategia Rozwoju Powiatu Poddębickiego 2020 jest','2MMIIAASSTTOO II GGMMIINNAA PPRRZZEEDDBBÓÓRRZZ',
            '4STRATEGIA ROZWOJU GMINY WIELUŃ 2025','1Strategia Rozwoju Gminy Miasta Bochnia na lata 2022-2030','Strategia Rozwoju Gminy Alwernia na lata 2022-2030, jest dokumentem, który określa',
            '3Strategia Rozwoju Gminy Trzebinia na lata 2014-2020','4STRATEGIA ROZWOJU GMINY SZCZUCIN 2014 - 2020',
            'Strategia Rozwoju Gminy Świątniki Górne na lata 2015-2020+ 2Wprowadzenie','2Strategia Rozwoju Gminy Czarny Dunajec na lata 2021-2030',
            '3Strategia Rozwoju Gminy Wolbrom','System wdrażania Strategii Rozwoju Gminy Zator na lata 2014-2022',
            'STRATEGIA ROZWOJU GMINY KOSZYCE NA LATA 20212026','GMINY NOWE BRZESKOSTRATEGIA ROZWOJU GMINY NOWE BRZESKO',
            'Rysunek','Strategia Rozwoju Gminy Radłów na lata 2021-2030 1Wprowadzenie','"Projekt: „Nowoczesne Zarządzanie Publiczne” o numerze POKL.05.02.01-00-025/12, realizowany w ramach Programu"',
            '3C I. W','Aktualne dokumenty SZJ znajdują się w sieci komputerowej w portalu pod adresem http://qsystem.nowysacz.pl Dokumenty w wersji 3',
            'Strona 4 z 99Strategia Rozwoju Gminy Wyśmierzyce na lata 2021-2030','Strona 5STRATEGIA ROZWOJU GMINY ŻELECHÓW NA LATA 2014 – 2020',
            'Strategia Rozwoju Gminy Miasta Gostynina jest dokumentem zawierającym długookresowy','Gminy Warka na lata 2005-2013 – jest sformułowanie syntetycznego dokumentu',
            'Strategia Rozwoju Gminy Kozienice na lata 2018-2030jest kierunkowo-programowym planem','Październik 2007 2Zaktualizowana Strategia Rozwoju Gminy Miejskiej Legionowo do 2015 roku',
            '2Strategia Rozwoju Gminy Solec nad Wisłą na lata 2007 - 2020','SPIS MAP','Strategia Zrównoważonego Rozwoju Gminy Halinów na lata 2016-2025',
            'Plan Rozwoju Lokalnego to dokument, który przedstaw ia zintegrowane i kompleksowe dzia łania','STRATEGIA ZRÓWNOWAŻONEGO ROZWOJU MIASTA I GMINY ZAKROCZYM DO 2020 ROKU',
            '3STRATEGIA ROZWOJU SPOŁECZNO-GOSPODARCZEGO MIASTA OTWOCKA','3STRATEGIA ROZWOJU GMINY SOCHOCIN NA LATA 2016-2020 z perspektywą do roku 2025',
            '3Strategia ZrównowaŜonego Rozwoju Miasta Piastowa do 2020 roku','Strona | 1STRATEGIA ROZWOJU GMINY MIASTO PIONKI NA LATA 2016-2022',
            'Spis tablic','W myśl rozwiązań ustawowych Władze Gminy Łomianki! Zobowiązane są do działań na','E O2OCHAOC 4724477 UE FTSWORZKL8. Podpeay son',
            'Przesłanki aktualizacji Strategii Rozwoju Gminy Zwoleń','Rozwój regionalny i lokalny ma szczególne znaczenie dla procesów wzrostu gospodarczego',
            'Opracowanie: JODKA Consulting 3Szanowni Państwo','SPIS, TABEL','Gmina Głubczyce położona w południowo-zachodniej części Polski na terenie województwa',
            '3Strategia Rozwoju Gminy Byczyna na lata 2015-2020','3ZESPÓŁ DS. OPRACOWANIA STRATEGII ROZWOJU GMINY KORFANTÓW',
            '"Zabytkowy kościół pw. św. Anny, oleskie masło i śmietankówka,','1Strategia Rozwoju Gminy Tułowice na lata: 2022-2030',
            'Autor fotografii na okładce: Damian Wojtowicz','2Strategia Rozwoju Gminy Kołaczyce do 2030 roku',
            'Strategia Rozwoju Gminy Przecław na lata 2016-2026 jest opracowaniem, którego treść','JKU Konsulting Sp. z o.o. 6„Strategia Zrównoważonego Rozwoju Gminy i Miasta Nisko na lata 2014-2020',
            '3Strategia Rozwoju 2021-2030','3Strategia Rozwoju Gminy Głogów','3Syntetyczne podsumowanie dokumentu SRMT 2021-2027',
            'Gmina Łapy obejmuje powierzchnię','3Plan Rozwoju Lokalnego Gminy Szczuczyn na lata 2008 - 2015','Strategia Rozwoju Gminy Jedwabne na lata 2021–2030',
            '2PLAN ROZWOJU LOKALNEGO GMINY NOWOGRÓD NA LATA 2016– 2022','Skuteczność w wykorzystywaniu funduszy strukturalnych Unii Europejskiej',
            'Niniejszy „Raport o Stanie Gminy” stanowi pierwszy etap prac nad Strategią','Proces prac nad Strategią Rozwoju Gminy Ciechanowiec na lata 2021 - 2030',
            'Głównym celem opracowania jest nakreślenie kierunków i priorytetów rozwoju gminy Czyżew na lata 2009-2017','Wykaz stosowanych skrótów',
            '2Stan gminy został oceniony według dostępnych, najnowszych danych. Podstawę do analiz','Strategia Rozwoju Gminy Nowy Dwór Gdański na lata 2021-2030',
            'Podjęcie prac nad Strategią ZrównowaŜonego Rozwoju Miasta Jastarni','Strategia rozwoju gminy jest zestawieniem celów, potrzeb i sposobów ich',
            'Id: . Uchwalony Strona 4WSTĘP','Sopot z Gdańskiem i Gdynią tworzą rdzeń Obszaru Metropolitalnego Gdańsk-Gdynia-Sopot',
            'Strategia Rozwoju Gminy Sośnicowice na lata 2013 – 2025','Obserwacja procesów społecznych, gospodarczych i kulturowych, których Samorząd',
            '3Strategia Rozwiązywania Problemów Społecznych Miasta Imielin','Dokument „Strategia Rozwoju Chorzowa do 2030 roku” stanowi rezultat części projektu pod nazwą',
            '4Przedkładana Państwu Strategia Zintegrowanego i Zrównoważonego Rozwoju Jaworzna','Strona 4 z 157STRATEGIA ROZWOJU GMINY WODZISŁAW NA LATA 2015-2025',
            'Zarządzanie gminą wiąże się z podejmowaniem wielu działań i decyzji ważnych dla','LISTA DOKUMENTÓW GMINY MORAWICA O ZNACZENIU STRATEGICZNYM',
            'STRATEGIA ROZWOJU GMINY IWANISKA NA LATA 2017-2025','Id: . Uchwalony Strona 4Strategi a Rozwoju Gminy Połaniec na lata 2021 - 2030 ',
            'Z wielką przyjemnością przedstawiam Państwu „Strategię Rozwoju Miasta Braniewa do','Strategia Zrównoważonego Rozwoju Miasta i Gminy Pieniężno 31. Proces tworzenia strategii i jego uczestnicy.',
            'Luty 2016r. 2Program Rozwoju Miasta i Gminy Tolkmicko na lata 2016-2025','Miasto Ełk i Gmina Ełk współpracują w ramach Miejskiego Obszaru Funkcjonalnego Ełku',
            'Miasta i Gminy Ryn na lata 2021–2031','Zarządzanie Gminą wiąże się z podejmowaniem szeregu działań i decyzji ważnych dla',
            'la V 2. V a Va Va Va WAGA WW. WE WZA U WAGA CU','2Strategia rozwoju społeczno-gospodarczego gminy Mrągowo do 2025 r. – raport z realizacji',
            'Strona 2Strategia Rozwoju Gminy Mikołajki do roku 2025 – PROJEKT','EKOTOP dr inż. Roman Sobczyk',
            'Dotychczas obowiązującym dokumentem strategicznym w Gminie Czerniejewo była uchwalona','2Strategia Rozwoju Gminy Gostyń na lata 2022-2030',
            'Gmina Żerków leży w województwie wielkopolskim w powiecie jarocińskim i sąsiaduje','W zastosowanym ujęciu proces planowania strategicznego obejmuje sześć następujących',
            '3Strategia Rozwoju Gminy Zduny na lata 2016-2023','4PLAN ROZWOJU LOKALNEGO GMINY RYDZYNA NA LATA 2016-2022',
            '3Mosina 2020+ [STRATEGIA ROZWOJU GMINY MOSINA]', 'Strategia Rozwoju Miast i Gminy Ostroróg to dokument określający strategię działań',
            'Prezentowane w diagnozie dane dotyczą obszaru gminy miejsko-wiejskiej Września','Strategia rozwoju Konina powstaje jako aktualizacja już istniejącego dokumentu z 2007',
            '3 | St ro naStrategia Rozwoju Gminy Choszczno na lata 2021-2030', 'Plan Rozwoju Lokalnego Gminy Drawno to dokument zawierający najważniejsze',
            'Strategia Rozwoju Gminy Czaplinek na lata 2014 ‐ 2020I. Raport o stanie Gminy Czaplinek','2 | St ro naStrategia Rozwoju Gminy Złocieniec na lata 2023-2028',
            '2Strategia Goleniów 2014-2023','Strategia Rozwoju Gminy Maszewo do 2030 roku','Spis map, rysunków, tabel i wykresów',
            '2Strategia/ Program Rozwoju Gminy Chojna','2Strategia Rozwoju Społeczno – Gospodarczego Gminy Mieszkowice na lata 2014 - 2024',
            ',,Strategia Rozwoju Gminy Dziwnów na lata 2016- 2025” jest dokumentem określającym"','4Strategia Rozwoju Gminy Kamień Pomorski na lata 2023-2030',
            'Przyszłość ma wiele nazw. Dla słabych oznacza nieuchronność, dla bojaźliwych nieznane','3Strategia Rozwoju Gminy Mielno na lata 2014-2022',
            'Id: . PodpisanyStrategia Rozwoju Gminy Dębno na lata 2023-2030','XVIII.	Zadania do realizacji w latach 2007-2020','Podstawą rozwoju gminy Lipiany jest strategia, która określa, misję, cele i kierunki działania',
            '2Strategia rozwoju gminy Pyrzyce na lata 2015-2025','56GŁÓWNE WNIOSKI Z DIAGNOZY','Strategia rozwoju Gminy Biały Bór na lata 2019 -2025 jest elementem polityki rozwoju',
            'Indeks rysunków',' Analiza otoczenia oraz stanu gminy i miasta Mirosławiec','PROGRAM 1 (PR-1) ZRÓWNOWAŻONY ROZWÓJ FUNKCJI',
            'W 2008 roku Rada Miasta Szczecina (RM) podjęła uchwałę1 o aktualizacji Strategii Rozwoju Szczecina, przyjętej',
            'Strategia Rozwoju Gminy Dziwnów na lata 2016- 2025” jest dokumentem określającym','Planowanie i zarz ądzanie strategiczne rozwoju jednostek terytorialnych jest',
            'Strategia rozwoju gminy jest drugim obok studium uwarunkowań i kierunków zagospodarowania przestrzennego gminy dokumentem wyznaczającym kierunki rozwoju obszaru',
            'Zabytkowy kościół pw. św. Anny, oleskie masło i śmietankówka, Rynek, stadion i Promenada oraz','Projekt: „Nowoczesne Zarządzanie Publiczne” o numerze POKL.05.02.01-00-025/12, realizowany w ramach Programu',
            '4Strategia rozwoju Gminy Rzepin'
        ]
        extracted_text = extract_section(text, start_keywords, end_keywords)
        if extracted_text:
            keywords = ['cel', 'celów', 'wizja', 'misja']
            processed_text_with_pages = process_text_with_pages(extracted_text, keywords, table_of_contents_pages=(1, 3))
            results = []
            for sentence, start_page, end_page in processed_text_with_pages:
                if pdf:
                    extracted_text = extract_pages(pdf, [(start_page, end_page)])
                else:
                    extracted_text = ''
                results.append({
                    'file_name': os.path.basename(file_path),
                    'sentence': sentence,
                    'start_page': start_page,
                    'end_page': end_page,
                    'extracted_text': extracted_text
                })
            return results
    return None

def process_directory(directory_path):
    results = []
    for root, _, files in os.walk(directory_path):
        for file in files:
            file_path = os.path.join(root, file)
            result = process_single_file(file_path)
            if result:
                results.extend(result)
    return results

def main():
    directory_path = 'D:/AA Praktyki/strategy' 
    results = process_directory(directory_path)
    
    if results:
        for result in results:
            print(f"Plik: {result['file_name']}")
            print(f"Nagłówek: {result['sentence']}")
            print(f"Strony: {result['start_page']}-{result['end_page']}")
            print(result['extracted_text'])
            print("\n" + "="*80 + "\n")
        
        df = pd.DataFrame(results)
        df.to_csv('extracted_texts.csv', index=False)
        print(f"Wyniki zapisane do pliku extracted_texts.csv")
    else:
        print("Nie znaleziono żadnych wyników.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import pytesseract
import spacy
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import re

nlp = spacy.load("pl_core_news_lg")

def remove_hyphenation(text):
    if isinstance(text, str): 
        text = re.sub(r'(\w+)-\n(\w+)', r'\1\2', text)
        text = text.replace('\n', ' ')
    else:
        text = '' 
    return text

def clean_text(text):
    if isinstance(text, str): 
        text = re.sub(r'[^\w\s]', '', text)
        text = re.sub(r'\d+', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
    else:
        text = ''  
    return text

def lemmatize_text(text):
    if isinstance(text, str): 
        try:
            doc = nlp(text)
            lemmas = [token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct]
            return ' '.join(lemmas), lemmas
        except Exception:
            return "", []
    return "", []

csv_path = 'C:/Users/angel/extracted_texts.csv'  
df_extracted = pd.read_csv(csv_path)
xlsx_path = 'D:/AA Praktyki/excel/branża_nazwa_rozsz.xlsx'
df = pd.read_excel(xlsx_path, engine='openpyxl')

def lemmatize_nazwa_column(df):
    df['nazwa_lematyzowana'], df['lemmas'] = zip(*df['nazwa'].apply(lambda x: lemmatize_text(x) if isinstance(x, str) else (x, [])))
    return df

df = lemmatize_nazwa_column(df)
all_results = []
extracted_words = []
lemmatized_words = []

for index, row in df_extracted.iterrows():
    file_name = row['file_name']
    extracted_text = row['extracted_text']
    
    cleaned_text = remove_hyphenation(extracted_text)
    cleaned_text = clean_text(cleaned_text)
    lemmatized_text, lemmas = lemmatize_text(cleaned_text)
    
    if lemmatized_text:
        vectorizer = TfidfVectorizer()
        tfidf_matrix = vectorizer.fit_transform([lemmatized_text] + df['nazwa_lematyzowana'].tolist())
        similarity_matrix = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])
        temp_df = pd.DataFrame({'file_name': file_name, 'branża': df['branża'], 'match_score': similarity_matrix[0]})
        data = temp_df[['match_score']].values
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(data)
        dbscan = DBSCAN(eps=0.5, min_samples=2)
        clusters = dbscan.fit_predict(data_scaled)
        temp_df['cluster'] = clusters
        cluster_mean_scores = temp_df.groupby(['file_name', 'cluster'])['match_score'].mean().reset_index()
        cluster_mean_scores.columns = ['file_name', 'cluster', 'mean_match_score']
        min_mean_scores = cluster_mean_scores.groupby('file_name')['mean_match_score'].transform('min')
        cluster_mean_scores['is_min'] = cluster_mean_scores['mean_match_score'] == min_mean_scores
        filtered_clusters = cluster_mean_scores[~cluster_mean_scores['is_min']]
        final_results = temp_df.merge(filtered_clusters[['file_name', 'cluster']], on=['file_name', 'cluster'], how='inner')
        all_results.append(final_results)
        extracted_words.extend([(file_name, word) for word in cleaned_text.split()])
        lemmatized_words.extend([(file_name, lemma) for lemma in lemmas])
if all_results: 
    all_results_df = pd.concat(all_results, ignore_index=True)
    extracted_words_df = pd.DataFrame(extracted_words, columns=['file_name', 'word'])
    lemmatized_words_df = pd.DataFrame(lemmatized_words, columns=['file_name', 'lemma'])
    extracted_words_df.to_csv('extracted_words.csv', index=False)
    lemmatized_words_df.to_csv('lemmatized_words.csv', index=False)
    branche_count = all_results_df.groupby('file_name')['branża'].nunique().reset_index(name='branche_count')

    plt.figure(figsize=(10, 6))
    plt.hist(branche_count['branche_count'], bins=20, edgecolor='k', alpha=0.7)
    plt.xlabel('Liczba branż')
    plt.ylabel('Liczba gmin')
    plt.title('Histogram liczby branż w gminach po filtracji')
    plt.grid(axis='y', alpha=0.75)
    plt.show()

    all_results_df.to_csv('filtered_clusters_results_all_files.csv', index=False)
else:
    print("Brak wyników do przetworzenia.")

```
