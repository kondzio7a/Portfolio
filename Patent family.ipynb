# Patent family

```python
import pandas as pd 
df_patents = pd.read_csv('D:/AA Praktyki/df_patents9020.csv')
df_patents.head()
patent_array=df_patents['id'].unique()
patent_array=patent_array[20168:]

```

```python
import pandas as pd
import re
import time
import random
from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_argument("--headless") 
driver = webdriver.Chrome(options=options)
def get_patent_priority_dataframe(patent_number):
    try:
        formatted_patent = patent_number.replace("P.", "PL")
        url = f"https://patents.google.com/patent/{formatted_patent}/en"
        driver.get(url)
        time.sleep(random.uniform(3, 7)) 
        visible_text = driver.execute_script("return document.body.innerText")
        matches = re.findall(r"Priority to\s+([^\s]+)", visible_text)
        if matches:
            df = pd.DataFrame(matches, columns=["priority_code"])
            df.insert(0, "id", patent_number)
            return df
        else:
            print(f"Brak wyników dla patentu: {patent_number}")
            return pd.DataFrame(columns=["id", "priority_code"])

    except Exception as e:
        print(f"Błąd dla patentu {patent_number}: {e}")
        return pd.DataFrame(columns=["id", "priority_code"])
    
all_priority_codes = pd.DataFrame(columns=["id", "priority_code"])

output_file = "D:/AA Praktyki/patent_priorities4.csv"

for idx, patent_number in enumerate(patent_array, start=1):
    print(f"Przetwarzanie {idx}/{len(patent_array)}: {patent_number}")

    df = get_patent_priority_dataframe(patent_number)
    all_priority_codes = pd.concat([all_priority_codes, df], ignore_index=True)

    if idx % 10 == 0 or idx == len(patent_array): 
        all_priority_codes.to_csv(output_file, index=False)
        print(f"Wyniki zapisano do pliku '{output_file}' po przetworzeniu {idx} patentów.")

    time.sleep(random.uniform(5, 15)) 

# Finalne zapisanie wyników
all_priority_codes.to_csv(output_file, index=False)
print(f"Wszystkie wyniki zostały zapisane do pliku '{output_file}'.")

# Zamknięcie przeglądarki
driver.quit()
```
