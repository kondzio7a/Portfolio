```python
import requests
import os
import json

# Ustawienia
krs = "0000339850" 
rejestr = "P" 
folder_docelowy = "D:/KRS"  
nazwa_pliku = f"{krs}_krs.json" 

# URL API
url = f"https://api-krs.ms.gov.pl/api/krs/OdpisPelny/{krs}?rejestr={rejestr}&format=json"
os.makedirs(folder_docelowy, exist_ok=True)
sciezka_pliku = os.path.join(folder_docelowy, nazwa_pliku)
response = requests.get(url)
if response.status_code == 200:
    data = response.json()
    with open(sciezka_pliku, "w", encoding="utf-8") as plik:
        json.dump(data, plik, ensure_ascii=False, indent=4)
    
    print(f"Dane zostały zapisane w pliku: {sciezka_pliku}")
else:
    print(f"Błąd podczas pobierania danych: {response.status_code}")
```

```python
import pandas as pd
import requests
import os
import json

# Ścieżka do pliku CSV
sciezka_csv = "D:/AA Praktyki/debug_krs_numbers.csv" 
folder_docelowy = "D:/KRS" 
rejestr = "P" 
df = pd.read_csv(sciezka_csv, dtype={'krs': str})
os.makedirs(folder_docelowy, exist_ok=True)
for index, row in df.iterrows():
    krs = str(row['krs']).strip() 
    if not krs.isdigit():
        print(f"Nieprawidłowy numer KRS dla ID {row['id']}: {krs}")
        continue
    
    krs = krs.zfill(10)
    id_ = row['id']
    url = f"https://api-krs.ms.gov.pl/api/krs/OdpisPelny/{krs}?rejestr={rejestr}&format=json"
    nazwa_pliku = f"{id_}_{krs}.json" 
    sciezka_pliku = os.path.join(folder_docelowy, nazwa_pliku)
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        with open(sciezka_pliku, "w", encoding="utf-8") as plik:
            json.dump(data, plik, ensure_ascii=False, indent=4)
        
        print(f"Pobrano dane i zapisano w pliku: {sciezka_pliku}")
    else:
        print(f"Błąd podczas pobierania danych dla ID {id_}, KRS {krs}: {response.status_code}")
```
